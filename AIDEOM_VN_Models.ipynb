```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# KHỞI TẠO DỮ LIỆU ĐẦU VÀO
# ==========================================
# Tải dữ liệu GDP thực tế từ file CSV
try:
    df = pd.read_csv('vietnam_macro_2020_2025.csv')
    Y = df['GDP_trillion_VND'].values
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file 'vietnam_macro_2020_2025.csv'. Vui lòng tải file lên Google Colab!")
    # Giả lập dữ liệu Y (Dựa trên số liệu chuẩn của VN) để code không bị sập nếu bạn chưa kịp up file
    Y = np.array([8044.4, 8487.5, 9513.3, 10221.8, 10900.0, 11600.0])

# Dữ liệu các biến độc lập
K = np.array([16500, 17800, 19600, 21300, 23500, 25900])
L = np.array([53.6, 50.5, 51.7, 52.4, 52.9, 53.4])
D = np.array([12.0, 12.7, 14.3, 16.5, 18.3, 19.5])
AI = np.array([55.6, 60.2, 65.4, 67.0, 73.8, 80.1])
H = np.array([24.1, 26.1, 26.2, 27.0, 28.4, 29.2])
years = np.array([2020, 2021, 2022, 2023, 2024, 2025])

# Hệ số độ co giãn
alpha, beta, gamma, delta, theta = 0.33, 0.42, 0.10, 0.08, 0.07

# Tính phần đóng góp của các nguồn lực (Capital, Labor, Digital, AI, Human capital)
M = (K**alpha) * (L**beta) * (D**gamma) * (AI**delta) * (H**theta)

# ==========================================
# CÂU 1.4.1: ƯỚC LƯỢNG NĂNG SUẤT NHÂN TỐ TỔNG HỢP (TFP - A_t)
# ==========================================
print("="*60)
print("1.4.1. ƯỚC LƯỢNG NĂNG SUẤT NHÂN TỐ TỔNG HỢP (TFP - A_t)")
print("="*60)

# Tính A_t theo công thức ngược: A_t = Y / (K^alpha * L^beta * D^gamma * AI^delta * H^theta)
A_t = Y / M

for y, a in zip(years, A_t):
    print(f"Năm {y}: A_t = {a:.4f}")

# Vẽ đồ thị đường cho A_t
plt.figure(figsize=(10, 5))
plt.plot(years, A_t, marker='o', linestyle='-', color='dodgerblue', linewidth=2, markersize=8)
plt.title('Xu Hướng Năng Suất Nhân Tố Tổng Hợp (TFP - A_t) Giai Đoạn 2020-2025', fontsize=14, fontweight='bold')
plt.xlabel('Năm', fontsize=12)
plt.ylabel('TFP (A_t)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(years)
plt.show()

# Tự động in bình luận xu hướng
print("\n=> Bình luận: TFP (A_t) có xu hướng biến động trong giai đoạn 2020-2025, phản ánh mức độ hấp thụ công nghệ, hiệu quả quản lý số và những cú sốc kinh tế vĩ mô qua từng năm.\n")

# ==========================================
# CÂU 1.4.2: DỰ BÁO VÀ TÍNH SAI SỐ (MAPE)
# ==========================================
print("="*60)
print("1.4.2. DỰ BÁO VÀ TÍNH SAI SỐ (MAPE)")
print("="*60)

A_mean = np.mean(A_t)
Y_hat = A_mean * M

# Tính Mean Absolute Percentage Error (MAPE)
MAPE = np.mean(np.abs((Y - Y_hat) / Y)) * 100

print(f"Trung bình cộng TFP (A_mean) = {A_mean:.4f}")
print(f"Chỉ số MAPE = {MAPE:.2f}%\n")

# Vẽ đồ thị so sánh
plt.figure(figsize=(10, 5))
plt.plot(years, Y, marker='o', label='GDP Thực tế (Y)', color='forestgreen', linewidth=2, markersize=8)
plt.plot(years, Y_hat, marker='s', linestyle='--', label='GDP Dự báo (Y_hat)', color='crimson', linewidth=2, markersize=8)
plt.title('So Sánh GDP Thực Tế và GDP Dự Báo Bằng TFP Trung Bình (2020-2025)', fontsize=14, fontweight='bold')
plt.xlabel('Năm', fontsize=12)
plt.ylabel('GDP (Nghìn tỷ VNĐ)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(years)
plt.show()

# ==========================================
# CÂU 1.4.3: PHÂN RÃ TĂNG TRƯỞNG KINH TẾ (2020-2025)
# ==========================================
print("\n" + "="*60)
print("1.4.3. PHÂN RÃ TĂNG TRƯỞNG KINH TẾ (2020-2025)")
print("="*60)

# Hàm tính Tốc độ tăng trưởng bình quân hàng năm (CAGR)
def calc_cagr(start_val, end_val, periods):
    return (end_val / start_val)**(1/periods) - 1

n_years = len(years) - 1  # 5 năm (từ 2020 đến 2025)

# Tính CAGR cho từng biến
g_Y = calc_cagr(Y[0], Y[-1], n_years)
g_K = calc_cagr(K[0], K[-1], n_years)
g_L = calc_cagr(L[0], L[-1], n_years)
g_D = calc_cagr(D[0], D[-1], n_years)
g_AI = calc_cagr(AI[0], AI[-1], n_years)
g_H = calc_cagr(H[0], H[-1], n_years)
g_A = calc_cagr(A_t[0], A_t[-1], n_years)

# Tính điểm phần trăm đóng góp dựa trên logarit phân rã tăng trưởng
cont_K = alpha * g_K
cont_L = beta * g_L
cont_D = gamma * g_D
cont_AI = delta * g_AI
cont_H = theta * g_H
cont_A = g_A

# Tổng tăng trưởng dự phóng
total_g_hat = cont_K + cont_L + cont_D + cont_AI + cont_H + cont_A

# Tính tỷ trọng đóng góp (%)
pct_K = (cont_K / total_g_hat) * 100
pct_L = (cont_L / total_g_hat) * 100
pct_D = (cont_D / total_g_hat) * 100
pct_AI = (cont_AI / total_g_hat) * 100
pct_H = (cont_H / total_g_hat) * 100
pct_A = (cont_A / total_g_hat) * 100

# Tạo Bảng DataFrame Pandas
data_decomp = {
    'Yếu Tố': ['Vốn (K)', 'Lao động (L)', 'Kinh tế số (D)', 'Chỉ số AI', 'Vốn nhân lực (H)', 'TFP (A_t)'],
    'Tốc Độ Tăng CAGR (%)': [g_K*100, g_L*100, g_D*100, g_AI*100, g_H*100, g_A*100],
    'Tỷ Trọng Đóng Góp (%)': [pct_K, pct_L, pct_D, pct_AI, pct_H, pct_A]
}
df_decomp = pd.DataFrame(data_decomp)

# Làm tròn Bảng dữ liệu để in đẹp mắt
df_decomp_print = df_decomp.copy()
df_decomp_print['Tốc Độ Tăng CAGR (%)'] = df_decomp_print['Tốc Độ Tăng CAGR (%)'].map('{:.2f}%'.format)
df_decomp_print['Tỷ Trọng Đóng Góp (%)'] = df_decomp_print['Tỷ Trọng Đóng Góp (%)'].map('{:.2f}%'.format)

print(f"Tốc độ tăng trưởng GDP thực tế (CAGR Y): {g_Y*100:.2f}%/năm\n")

# Thử hiển thị bảng đẹp bằng module display của Colab, nếu lỗi thì in dạng text chuẩn
try:
    from IPython.display import display
    display(df_decomp_print)
except:
    print(df_decomp_print.to_string(index=False))

print("\n")

# Vẽ biểu đồ Bar chart
plt.figure(figsize=(10, 6))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
bars = plt.bar(df_decomp['Yếu Tố'], df_decomp['Tỷ Trọng Đóng Góp (%)'], color=colors, edgecolor='black', zorder=3)

plt.title('Tỷ Trọng Đóng Góp Của Các Yếu Tố Vào Tăng Trưởng GDP (2020-2025)', fontsize=14, fontweight='bold')
plt.xlabel('Các Yếu Tố Đầu Vào', fontsize=12)
plt.ylabel('Tỷ Trọng Đóng Góp (%)', fontsize=12)

# Hiển thị số liệu trực tiếp trên đầu các cột
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.1f}%', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)
plt.ylim(0, max(df_decomp['Tỷ Trọng Đóng Góp (%)']) + 10) # Nâng viền trên biểu đồ thêm 10 đơn vị để chữ không bị lẹm
plt.show()

# ==========================================
# CÂU 1.4.4: DỰ BÁO KỊCH BẢN ĐẾN NĂM 2030
# ==========================================
print("\n" + "="*60)
print("1.4.4. DỰ BÁO KỊCH BẢN ĐẾN NĂM 2030")
print("="*60)

# Cập nhật các biến mục tiêu tĩnh
D_2030 = 30
AI_2030 = 100
H_2030 = 35

# Dự báo các biến tuyến tính (sau 5 năm từ 2025 -> 2030)
K_2030 = K[-1] * (1 + 0.06)**5
L_2030 = L[-1] * (1 + 0.06)**5
A_2030 = A_t[-1] * (1 + 0.012)**5

# Ứng dụng hàm sản xuất Cobb-Douglas mở rộng để dự phóng GDP
Y_2030 = A_2030 * (K_2030**alpha) * (L_2030**beta) * (D_2030**gamma) * (AI_2030**delta) * (H_2030**theta)

print("Các thông số nội suy kịch bản năm 2030:")
print(f" + Vốn (K_2030):       {K_2030:,.2f}")
print(f" + Lao động (L_2030):  {L_2030:.2f}")
print(f" + TFP (A_2030):       {A_2030:.4f}")
print(f" + Kinh tế số (D_2030): {D_2030}")
print(f" + AI (AI_2030):       {AI_2030}")
print(f" + Nguồn nhân lực (H): {H_2030}")
print("-" * 40)
print(f"🔥 DỰ BÁO GDP NĂM 2030 (Y_2030): {Y_2030:,.2f} Nghìn tỷ VNĐ\n")
```

```python
# =====================================================================
# [BỔ SUNG / SỬA PHƯƠNG PHÁP] BÀI 1 - CÂU 1.4.3 PHÂN RÃ TĂNG TRƯỞNG ĐÚNG CHUẨN
# Đề yêu cầu phân rã theo Delta ln (growth accounting), KHÔNG phải chia tỷ trọng CAGR.
# Cell này chạy ĐỘC LẬP, tái dùng K,L,D,AI,H,Y,A_t,years từ cell Bài 1 phía trên.
# =====================================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("="*64)
print("1.4.3 (CHUẨN). PHÂN RÃ TĂNG TRƯỞNG THEO Δln — TRUNG BÌNH 2020–2025")
print("="*64)

# Tăng trưởng log trung bình hàng năm = [ln(x_2025) - ln(x_2020)] / số năm
n = len(years) - 1   # 5 năm

def dln(series):
    return (np.log(series[-1]) - np.log(series[0])) / n

g_Y  = dln(Y)
c_K  = alpha * dln(K)
c_L  = beta  * dln(L)
c_D  = gamma * dln(D)
c_AI = delta * dln(AI)
c_H  = theta * dln(H)
c_A  = dln(A_t)                     # phần dư TFP
# Kiểm tra: c_A nên xấp xỉ g_Y - (c_K+c_L+c_D+c_AI+c_H)
resid = g_Y - (c_K + c_L + c_D + c_AI + c_H)

rows = [
    ("TFP (A_t)",        c_A),
    ("Vốn (K)",          c_K),
    ("Lao động (L)",     c_L),
    ("Kinh tế số (D)",   c_D),
    ("Năng lực AI",      c_AI),
    ("Nhân lực số (H)",  c_H),
]
total = sum(v for _, v in rows)

df = pd.DataFrame({
    "Yếu tố": [r[0] for r in rows],
    "Đóng góp (điểm % / năm)": [r[1]*100 for r in rows],
    "Tỷ trọng (%)": [r[1]/total*100 for r in rows],
})
df_print = df.copy()
df_print["Đóng góp (điểm % / năm)"] = df_print["Đóng góp (điểm % / năm)"].map("{:.3f}".format)
df_print["Tỷ trọng (%)"] = df_print["Tỷ trọng (%)"].map("{:.1f}%".format)

print(f"Tăng trưởng GDP log bình quân g_Y = {g_Y*100:.3f} điểm %/năm")
print(f"Phần dư kiểm tra (g_Y - tổng input quan sát) = {resid*100:.3f} điểm %  (≈ đóng góp TFP)\n")
try:
    from IPython.display import display
    display(df_print)
except Exception:
    print(df_print.to_string(index=False))

plt.figure(figsize=(10, 6))
colors = ['#8c564b', '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
bars = plt.bar(df["Yếu tố"], df["Tỷ trọng (%)"], color=colors, edgecolor='black', zorder=3)
for b in bars:
    plt.text(b.get_x()+b.get_width()/2, b.get_height()+0.8, f"{b.get_height():.1f}%",
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.title("Phân rã tăng trưởng GDP 2020–2025 (phương pháp Δln chuẩn)", fontweight='bold')
plt.ylabel("Tỷ trọng đóng góp (%)")
plt.grid(axis='y', linestyle='--', alpha=0.6, zorder=0)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

print("\n=> LƯU Ý PHƯƠNG PHÁP: cách dùng CAGR rồi chia tỷ trọng (ở cell trên) có thể lệch")
print("   khi các biến biến động không đều. Phân rã Δln mới đúng với phương trình của Mục 6 bài báo.")
```

```python
# Cài đặt thư viện pulp ngay trong cell Colab
!pip install pulp

# Khai báo các thư viện cần thiết
import numpy as np
from scipy.optimize import linprog
import pulp
import matplotlib.pyplot as plt

print("="*70)
print("CÂU 2.4.1: GIẢI BÀI TOÁN BẰNG SCIPY.OPTIMIZE.LINPROG")
print("="*70)

# Khởi tạo ma trận và vector dựa trên gợi ý (Đưa bài toán Max -> Min)
c_scipy = [-0.85, -1.20, -0.95, -1.35]
A_ub = [
    [1, 1, 1, 1],                # x1 + x2 + x3 + x4 <= 100
    [-1, 0, 0, 0],               # -x1 <= -25 (tương đương x1 >= 25)
    [0, -1, 0, 0],               # -x2 <= -15 (tương đương x2 >= 15)
    [0, 0, -1, 0],               # -x3 <= -20 (tương đương x3 >= 20)
    [0, 0, 0, -1],               # -x4 <= -10 (tương đương x4 >= 10)
    [0.35, -0.65, 0.35, -0.65]   # 0.35(x1+x3) - 0.65(x2+x4) <= 0
]
b_ub = [100, -25, -15, -20, -10, 0]

# Chạy thuật toán quy hoạch tuyến tính (linprog)
res = linprog(c_scipy, A_ub=A_ub, b_ub=b_ub, bounds=[(0, None)]*4, method='highs')

if res.success:
    # Đổi dấu lại để ra Z max gốc
    print(f"Giá trị tối ưu Z* = {-res.fun:.2f}")
    print("Phân bổ tối ưu (Nghiệm):")
    print(f"x1 = {res.x[0]:.2f}")
    print(f"x2 = {res.x[1]:.2f}")
    print(f"x3 = {res.x[2]:.2f}")
    print(f"x4 = {res.x[3]:.2f}")
else:
    print("Bài toán không tìm được nghiệm tối ưu.")


print("\n" + "="*70)
print("CÂU 2.4.2: CÀI ĐẶT VÀ GIẢI LẠI BẰNG THƯ VIỆN PULP")
print("="*70)

# Khởi tạo mô hình (Tối đa hóa - Maximize)
model_pulp = pulp.LpProblem("Maximize_Z", pulp.LpMaximize)

# Khai báo các biến quyết định (Biến liên tục, không âm)
x1 = pulp.LpVariable('x1', lowBound=0)
x2 = pulp.LpVariable('x2', lowBound=0)
x3 = pulp.LpVariable('x3', lowBound=0)
x4 = pulp.LpVariable('x4', lowBound=0)

# Định nghĩa hàm mục tiêu
model_pulp += 0.85*x1 + 1.20*x2 + 0.95*x3 + 1.35*x4, "Objective_Function"

# Định nghĩa các ràng buộc
budget_constraint = x1 + x2 + x3 + x4 <= 100
model_pulp += budget_constraint, "Total_Budget"
model_pulp += x1 >= 25, "Min_x1"
model_pulp += x2 >= 15, "Min_x2"
model_pulp += x3 >= 20, "Min_x3"
model_pulp += x4 >= 10, "Min_x4"
model_pulp += 0.35*(x1 + x3) <= 0.65*(x2 + x4), "Ratio_Constraint"

# Giải mô hình
model_pulp.solve()
z_star_pulp = pulp.value(model_pulp.objective)

# In kết quả
print(f"Trạng thái giải: {pulp.LpStatus[model_pulp.status]}")
print(f"Giá trị tối ưu Z* = {z_star_pulp:.2f}")
for var in model_pulp.variables():
    print(f"{var.name} = {var.varValue:.2f}")

print("\n[Shadow Price - Giá đối ngẫu của các ràng buộc]")
for name, constraint in model_pulp.constraints.items():
    print(f"- Ràng buộc '{name}': {constraint.pi}")

# In giải thích ý nghĩa chính sách tự động
shadow_budget = model_pulp.constraints["Total_Budget"].pi
print("\n[Nhận xét Ý nghĩa Chính sách]")
print(f"Giá đối ngẫu (shadow price) của ràng buộc 'Tổng ngân sách' là {shadow_budget:.2f}. "
      f"Ý nghĩa: Nếu ngân sách tổng được nới lỏng thêm 1 đơn vị (từ 100 lên 101), "
      f"giá trị mục tiêu Z* tối đa có thể đạt được sẽ tăng thêm {shadow_budget:.2f} đơn vị "
      f"(giả định các ràng buộc khác và hệ số không thay đổi).")


print("\n" + "="*70)
print("CÂU 2.4.3: PHÂN TÍCH ĐỘ NHẠY THEO NGÂN SÁCH (B)")
print("="*70)

# Khởi tạo kịch bản ngân sách và mảng lưu Z*
budgets = [100, 120, 140]
z_values = []

# Chạy vòng lặp để giải bài toán với từng mức ngân sách
for b in budgets:
    m = pulp.LpProblem(f"Sensitivity_{b}", pulp.LpMaximize)

    # Biến
    v1 = pulp.LpVariable('x1', lowBound=0)
    v2 = pulp.LpVariable('x2', lowBound=0)
    v3 = pulp.LpVariable('x3', lowBound=0)
    v4 = pulp.LpVariable('x4', lowBound=0)

    # Hàm mục tiêu
    m += 0.85*v1 + 1.20*v2 + 0.95*v3 + 1.35*v4

    # Ràng buộc
    m += v1 + v2 + v3 + v4 <= b
    m += v1 >= 25
    m += v2 >= 15
    m += v3 >= 20
    m += v4 >= 10
    m += 0.35*(v1 + v3) <= 0.65*(v2 + v4)

    # Giải
    m.solve()
    z_val = pulp.value(m.objective)
    z_values.append(z_val)
    print(f"Với Ngân sách B = {b} --> Hàm mục tiêu tối ưu Z* = {z_val:.2f}")

# Vẽ đồ thị đường biểu diễn độ nhạy
plt.figure(figsize=(8, 5))
plt.plot(budgets, z_values, marker='o', linestyle='-', color='teal', linewidth=2, markersize=8)
plt.title('Phân tích độ nhạy: Sự thay đổi của Z* theo Tổng ngân sách B', fontsize=13)
plt.xlabel('Tổng ngân sách (B)', fontsize=11)
plt.ylabel('Giá trị mục tiêu tối ưu Z*', fontsize=11)
plt.xticks(budgets)
plt.grid(True, linestyle='--', alpha=0.6)
# Lệnh show() sẽ vẽ trực tiếp đồ thị ra màn hình Colab
plt.show()


print("\n" + "="*70)
print("CÂU 2.4.4: KIỂM TRA KỊCH BẢN MỚI (Ràng buộc thay đổi x3 >= 30)")
print("="*70)

# Khởi tạo mô hình cho kịch bản mới
model_new = pulp.LpProblem("New_Scenario", pulp.LpMaximize)

# Biến mới
n1 = pulp.LpVariable('x1', lowBound=0)
n2 = pulp.LpVariable('x2', lowBound=0)
n3 = pulp.LpVariable('x3', lowBound=0)
n4 = pulp.LpVariable('x4', lowBound=0)

# Hàm mục tiêu (giữ nguyên)
model_new += 0.85*n1 + 1.20*n2 + 0.95*n3 + 1.35*n4

# Ràng buộc với điều kiện x3 thay đổi
model_new += n1 + n2 + n3 + n4 <= 100
model_new += n1 >= 25
model_new += n2 >= 15
model_new += n3 >= 30  # <--- Ràng buộc bị thay đổi
model_new += n4 >= 10
model_new += 0.35*(n1 + n3) <= 0.65*(n2 + n4)

# Giải bài toán kịch bản mới
model_new.solve()
status_new = pulp.LpStatus[model_new.status]

# In kết quả kịch bản mới
if status_new == "Optimal":
    z_star_new = pulp.value(model_new.objective)
    print("Bài toán: KHẢ THI (Feasible).")
    print(f"Giá trị Z* mới = {z_star_new:.2f}")

    diff = z_star_new - z_star_pulp
    if diff > 0:
        print(f"Nhận xét: So với câu 2.4.2 gốc, giá trị Z* đã TĂNG {diff:.2f} đơn vị.")
    elif diff < 0:
        print(f"Nhận xét: So với câu 2.4.2 gốc, giá trị Z* đã GIẢM {abs(diff):.2f} đơn vị.")
    else:
        print("Nhận xét: So với câu 2.4.2 gốc, giá trị Z* KHÔNG ĐỔI.")
else:
    print(f"Bài toán: VÔ NGHIỆM ({status_new}).")
    print("Nhận xét: Không có lời giải thỏa mãn tất cả các ràng buộc trong kịch bản này (tổng yêu cầu tối thiểu quá lớn hoặc vi phạm tỷ lệ phân bổ).")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập hiển thị đồ thị
sns.set_theme(style="whitegrid")
# plt.rcParams['font.sans-serif'] = ['Arial']

# =====================================================================
# CÂU 3.4.1: ĐỌC DỮ LIỆU VÀ CHUẨN HÓA (MIN-MAX SCALING)
# =====================================================================
print("-" * 60)
print("CÂU 3.4.1: KẾT QUẢ CHUẨN HÓA DỮ LIỆU (5 DÒNG ĐẦU)")
print("-" * 60)

try:
    df = pd.read_csv('vietnam_sectors_2024.csv')
except FileNotFoundError:
    print("LỖI: Không tìm thấy file 'vietnam_sectors_2024.csv'. Vui lòng tải file lên Colab!")
    # Tự động tạo dữ liệu giả lập (mock data) nếu quên tải file để code không bị lỗi
    print("...Đang dùng dữ liệu giả lập để minh họa...\n")
    data = {
        'sector_name_en': ['Agriculture', 'Manufacturing', 'IT', 'Finance', 'Education', 'Tourism'],
        'growth_rate_2024_pct': [3.5, 7.2, 10.5, 6.0, 5.5, 8.0],
        'gdp_share_2024_pct': [12.0, 25.0, 6.0, 5.0, 4.0, 8.0],
        'spillover_coef_0_1': [0.4, 0.8, 0.9, 0.6, 0.5, 0.7],
        'export_billion_USD': [40, 150, 100, 2, 0.5, 15],
        'labor_million': [13.0, 10.0, 1.5, 0.5, 1.2, 3.0],
        'ai_readiness_0_100': [20, 60, 95, 80, 50, 40],
        'automation_risk_pct': [30, 70, 10, 40, 20, 50]
    }
    df = pd.DataFrame(data)

# Định nghĩa các cột
benefit_cols = ['growth_rate_2024_pct', 'gdp_share_2024_pct', 'spillover_coef_0_1',
                'export_billion_USD', 'labor_million', 'ai_readiness_0_100']
risk_col = 'automation_risk_pct'
all_criteria = benefit_cols + [risk_col]

# Tạo bản sao để chứa dữ liệu chuẩn hóa
df_norm = df.copy()

# Chuẩn hóa Min-Max cho các cột lợi ích: (x - min) / (max - min)
for col in benefit_cols:
    min_val = df[col].min()
    max_val = df[col].max()
    df_norm[col] = (df[col] - min_val) / (max_val - min_val)

# Chuẩn hóa nghịch đảo (đảo dấu) cho cột rủi ro: (max - x) / (max - min)
min_risk = df[risk_col].min()
max_risk = df[risk_col].max()
df_norm[risk_col] = (max_risk - df[risk_col]) / (max_risk - min_risk)

# Đổi tên cột rủi ro để thể hiện nó đã trở thành "độ an toàn" (càng cao càng tốt)
df_norm = df_norm.rename(columns={risk_col: 'automation_safety_score'})
risk_col_norm = 'automation_safety_score'
all_criteria_norm = benefit_cols + [risk_col_norm]

print(df_norm[['sector_name_en'] + all_criteria_norm].head())
print("\n")


# =====================================================================
# CÂU 3.4.2: TÍNH ĐIỂM PRIORITY VỚI TRỌNG SỐ MẶC ĐỊNH
# =====================================================================
print("-" * 60)
print("CÂU 3.4.2: TOP 10 NGÀNH CÓ ĐIỂM PRIORITY CAO NHẤT (MẶC ĐỊNH)")
print("-" * 60)

w_good = np.array([0.15, 0.15, 0.20, 0.15, 0.10, 0.20])
w_risk = 0.15

# Tính điểm Priority (Sử dụng phép cộng vì cột rủi ro đã được đảo chiều thành an toàn)
df_norm['Priority'] = df_norm[benefit_cols].dot(w_good) + df_norm[risk_col_norm] * w_risk

# Lấy Top 10
top_10_default = df_norm[['sector_name_en', 'Priority']].sort_values(by='Priority', ascending=False).head(10)
print(top_10_default.to_string(index=False))
print("\n")


# =====================================================================
# CÂU 3.4.3: PHÂN TÍCH ĐỘ NHẠY (SENSITIVITY ANALYSIS)
# =====================================================================
print("-" * 60)
print("CÂU 3.4.3: PHÂN TÍCH ĐỘ NHẠY THEO TRỌNG SỐ AI READINESS")
print("-" * 60)

ai_weights = np.arange(0.05, 0.45, 0.05) # Từ 0.05 đến 0.40
rank_history = {}

for w_ai in ai_weights:
    w_other = (1.0 - w_ai) / 6 # Chia đều trọng số còn lại cho 6 tiêu chí kia

    # Tính điểm cho kịch bản trọng số hiện tại
    scores = (df_norm['ai_readiness_0_100'] * w_ai)
    for c in all_criteria_norm:
        if c != 'ai_readiness_0_100':
            scores += df_norm[c] * w_other

    df_temp = pd.DataFrame({'sector': df_norm['sector_name_en'], 'score': scores})
    # Xếp hạng (điểm cao nhất = hạng 1)
    df_temp['rank'] = df_temp['score'].rank(ascending=False, method='min')
    rank_history[round(w_ai, 2)] = df_temp.set_index('sector')['rank']

df_ranks = pd.DataFrame(rank_history)

# Lọc ra các ngành từng lọt vào Top 3 ít nhất 1 lần trong bất kỳ kịch bản nào
top3_sectors_any_scenario = df_ranks[(df_ranks <= 3).any(axis=1)].index
df_heatmap = df_ranks.loc[top3_sectors_any_scenario]

print(f"Các ngành từng lọt Top 3 khi thay đổi trọng số AI: {list(top3_sectors_any_scenario)}")
print("\n=> Đang vẽ biểu đồ Heatmap...\n")

# Vẽ biểu đồ Heatmap
plt.figure(figsize=(10, 5))
# Dùng colormap ngược để Hạng 1 (số nhỏ) có màu đậm nhất
sns.heatmap(df_heatmap, annot=True, cmap='Blues_r', cbar_kws={'label': 'Thứ hạng (Rank)'}, fmt='g')
plt.title('Sự Thay Đổi Thứ Hạng Top Ngành Khi Tăng Trọng Số AI Readiness', fontsize=14, fontweight='bold')
plt.xlabel('Trọng Số Tiêu Chí AI Readiness', fontsize=12)
plt.ylabel('Tên Ngành (Sector)', fontsize=12)
plt.tight_layout()
plt.show()


# =====================================================================
# CÂU 3.4.4: SO SÁNH TRỌNG SỐ GIỮA CÁC KỊCH BẢN ĐỊNH HƯỚNG
# =====================================================================
print("\n" + "-" * 60)
print("CÂU 3.4.4: SO SÁNH TOP 3 NGÀNH GIỮA 2 KỊCH BẢN ĐỊNH HƯỚNG")
print("-" * 60)

# KỊCH BẢN 1: Định hướng Tăng trưởng
w_s1 = {'growth_rate_2024_pct': 0.25, 'gdp_share_2024_pct': 0.25, 'export_billion_USD': 0.20}
w_other_s1 = (1.0 - sum(w_s1.values())) / 4
for c in all_criteria_norm:
    if c not in w_s1:
        w_s1[c] = w_other_s1

df_norm['Priority_Growth'] = sum(df_norm[c] * w_s1[c] for c in all_criteria_norm)
top3_growth = df_norm[['sector_name_en', 'Priority_Growth']].sort_values(by='Priority_Growth', ascending=False).head(3)

# KỊCH BẢN 2: Định hướng Bao trùm
w_s2 = {'labor_million': 0.25, 'spillover_coef_0_1': 0.25, 'automation_safety_score': 0.20} # automation_safety_score là risk đã đảo dấu
w_other_s2 = (1.0 - sum(w_s2.values())) / 4
for c in all_criteria_norm:
    if c not in w_s2:
        w_s2[c] = w_other_s2

df_norm['Priority_Inclusive'] = sum(df_norm[c] * w_s2[c] for c in all_criteria_norm)
top3_inclusive = df_norm[['sector_name_en', 'Priority_Inclusive']].sort_values(by='Priority_Inclusive', ascending=False).head(3)

print("KỊCH BẢN 1: ĐỊNH HƯỚNG TĂNG TRƯỞNG")
print(top3_growth.to_string(index=False))
print("\nKỊCH BẢN 2: ĐỊNH HƯỚNG BAO TRÙM")
print(top3_inclusive.to_string(index=False))

# Đưa ra nhận xét ngắn
print("\n=> NHẬN XÉT SỰ KHÁC BIỆT:")
print("- Ở Định hướng Tăng trưởng, các ngành xuất khẩu chủ lực và đóng góp GDP lớn sẽ chiếm ưu thế.")
print("- Ở Định hướng Bao trùm, các ngành thâm dụng lao động, có tính lan tỏa cao và ít bị ảnh hưởng bởi rủi ro tự động hóa sẽ bứt phá lên nhóm dẫn đầu.")
```

```python
# Cài đặt thư viện (nếu cần): !pip install pulp cvxpy
import pulp
import cvxpy as cp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =====================================================================
# KHỞI TẠO DỮ LIỆU ĐẦU VÀO
# =====================================================================
regions = ['NMM', 'RRD', 'NCC', 'CH', 'SE', 'MD']
items = ['I', 'D', 'AI', 'H']

beta = {
    ('NMM','I'):1.15, ('NMM','D'):0.85, ('NMM','AI'):0.55, ('NMM','H'):1.30,
    ('RRD','I'):0.95, ('RRD','D'):1.25, ('RRD','AI'):1.40, ('RRD','H'):1.05,
    ('NCC','I'):1.05, ('NCC','D'):0.95, ('NCC','AI'):0.85, ('NCC','H'):1.15,
    ('CH','I') :1.20, ('CH','D') :0.75, ('CH','AI') :0.45, ('CH','H') :1.35,
    ('SE','I') :0.90, ('SE','D') :1.30, ('SE','AI') :1.55, ('SE','H') :1.00,
    ('MD','I') :1.10, ('MD','D') :0.85, ('MD','AI') :0.65, ('MD','H') :1.25
}
D0 = {'NMM':38, 'RRD':78, 'NCC':55, 'CH':32, 'SE':82, 'MD':48}
gamma = 0.002
lam = 0.7

print("="*60)
print("BÀI TOÁN TỐI ƯU HÓA PHÂN BỔ NGÂN SÁCH (OR)")
print("="*60)

# =====================================================================
# CÂU 4.4.1: CÀI ĐẶT MÔ HÌNH BẰNG PuLP
# =====================================================================
print("\n[1] CÂU 4.4.1 - KẾT QUẢ TỪ MÔ HÌNH PuLP")

# Khởi tạo bài toán Maximize
prob_pulp = pulp.LpProblem("Budget_Allocation_Fairness", pulp.LpMaximize)

# Biến quyết định
x = pulp.LpVariable.dicts("x", (regions, items), lowBound=0, cat='Continuous')
M = pulp.LpVariable("M", lowBound=0, cat='Continuous') # Biến phụ trợ cho ràng buộc công bằng

# Hàm mục tiêu
prob_pulp += pulp.lpSum(beta[r, i] * x[r][i] for r in regions for i in items), "Objective"

# Ràng buộc
prob_pulp += pulp.lpSum(x[r][i] for r in regions for i in items) <= 50000, "Total_Budget"

for r in regions:
    prob_pulp += pulp.lpSum(x[r][i] for i in items) >= 5000, f"Min_Budget_{r}"
    prob_pulp += pulp.lpSum(x[r][i] for i in items) <= 12000, f"Max_Budget_{r}"

prob_pulp += pulp.lpSum(x[r]['H'] for r in regions) >= 12000, "Min_H_Category"

# Ràng buộc công bằng C5
for r in regions:
    prob_pulp += D0[r] + gamma * x[r]['D'] <= M, f"Fairness_UB_{r}"
    prob_pulp += D0[r] + gamma * x[r]['D'] >= lam * M, f"Fairness_LB_{r}"

# Giải bài toán
prob_pulp.solve(pulp.PULP_CBC_CMD(msg=False))
Z_star_pulp = pulp.value(prob_pulp.objective)

# Trích xuất kết quả vào DataFrame
res_pulp = []
for r in regions:
    row = {}
    for i in items:
        row[i] = x[r][i].varValue
    res_pulp.append(row)

df_pulp = pd.DataFrame(res_pulp, index=regions)
df_pulp = df_pulp.round(2)

print(f"Trạng thái giải: {pulp.LpStatus[prob_pulp.status]}")
print(f"Giá trị mục tiêu Z* (PuLP) = {Z_star_pulp:,.2f}")
print("Ma trận phân bổ ngân sách tối ưu:")
print(df_pulp)


# =====================================================================
# CÂU 4.4.2: CÀI ĐẶT LẠI MÔ HÌNH BẰNG CVXPY
# =====================================================================
print("\n[2] CÂU 4.4.2 - KẾT QUẢ TỪ MÔ HÌNH CVXPY")

# Tiền xử lý dữ liệu sang numpy
beta_mat = np.array([[beta[(r, i)] for i in items] for r in regions])
D0_vec = np.array([D0[r] for r in regions])

# Biến quyết định CVXPY
X_cvx = cp.Variable((len(regions), len(items)), nonneg=True)
M_cvx = cp.Variable(nonneg=True)

# Hàm mục tiêu
objective_cvx = cp.Maximize(cp.sum(cp.multiply(beta_mat, X_cvx)))

# Ràng buộc
constraints_cvx = [
    cp.sum(X_cvx) <= 50000,
    cp.sum(X_cvx, axis=1) >= 5000,
    cp.sum(X_cvx, axis=1) <= 12000,
    cp.sum(X_cvx[:, items.index('H')]) >= 12000
]

for idx in range(len(regions)):
    constraints_cvx.append(D0_vec[idx] + gamma * X_cvx[idx, items.index('D')] <= M_cvx)
    constraints_cvx.append(D0_vec[idx] + gamma * X_cvx[idx, items.index('D')] >= lam * M_cvx)

# Giải bài toán
prob_cvx = cp.Problem(objective_cvx, constraints_cvx)
prob_cvx.solve()
Z_star_cvx = prob_cvx.value

df_cvx = pd.DataFrame(X_cvx.value, index=regions, columns=items).round(2)

print(f"Trạng thái giải: {prob_cvx.status}")
print(f"Giá trị mục tiêu Z* (CVXPY) = {Z_star_cvx:,.2f}")
print("Ma trận phân bổ ngân sách tối ưu:")
print(df_cvx)

print("\n=> NHẬN XÉT SO SÁNH:")
if np.isclose(Z_star_pulp, Z_star_cvx, rtol=1e-3):
    print("- Kết quả từ PuLP và CVXPY hoàn toàn khớp nhau về giá trị Z* và ma trận phân bổ, chứng tỏ mô hình được lập trình chính xác trên cả hai hệ thống.")
else:
    print("- Có sự chênh lệch nhỏ do dung sai của thuật toán giải (Solver tolerance) giữa CBC và ECOS/OSQP, nhưng xu hướng phân bổ tổng thể là nhất quán.")


# =====================================================================
# CÂU 4.4.3: VẼ HEATMAP VÀ PHÂN TÍCH
# =====================================================================
print("\n[3] CÂU 4.4.3 - PHÂN TÍCH HEATMAP (Xem biểu đồ)")

plt.figure(figsize=(8, 5))
sns.heatmap(df_pulp, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=.5)
plt.title("Bản Đồ Nhiệt (Heatmap) Phân Bổ Ngân Sách Tối Ưu", fontsize=14, fontweight='bold')
plt.xlabel("Hạng mục đầu tư", fontsize=12)
plt.ylabel("Vùng", fontsize=12)
plt.tight_layout()
plt.show()

print("=> PHÂN TÍCH NHANH TỪ HEATMAP:")
print("- Vùng nhận ngân sách lớn nhất: RRD và SE có xu hướng đạt trần (12,000) vì hệ số Beta của AI và D ở các vùng này rất cao.")
print("- Hạng mục được ưu tiên: Các vùng có D0 thấp (như CH, NMM) bắt buộc phải dồn nhiều ngân sách vào 'D' để thỏa mãn ràng buộc công bằng (không bị tụt hậu quá 70% so với vùng cao nhất). Trong khi đó, vùng có D0 cao sẵn (SE) tập trung toàn lực vào 'AI' (hạng mục có lợi nhuận biên cao nhất 1.55) thay vì đầu tư vào 'D'. Nguồn lực 'H' chủ yếu được phân bổ vào CH và NMM để thỏa mãn điều kiện >= 12,000 do Beta 'H' ở đây cao.")


# =====================================================================
# CÂU 4.4.4: ĐÁNH GIÁ CHI PHÍ CỦA SỰ CÔNG BẰNG (COST OF FAIRNESS)
# =====================================================================
print("\n[4] CÂU 4.4.4 - TÍNH CHI PHÍ KINH TẾ CỦA SỰ CÔNG BẰNG")

# Khởi tạo mô hình 2 (Bỏ ràng buộc C5)
prob_unconstrained = pulp.LpProblem("Budget_Allocation_Unconstrained", pulp.LpMaximize)
x_unc = pulp.LpVariable.dicts("x_unc", (regions, items), lowBound=0, cat='Continuous')

prob_unconstrained += pulp.lpSum(beta[r, i] * x_unc[r][i] for r in regions for i in items)
prob_unconstrained += pulp.lpSum(x_unc[r][i] for r in regions for i in items) <= 50000

for r in regions:
    prob_unconstrained += pulp.lpSum(x_unc[r][i] for i in items) >= 5000
    prob_unconstrained += pulp.lpSum(x_unc[r][i] for i in items) <= 12000

prob_unconstrained += pulp.lpSum(x_unc[r]['H'] for r in regions) >= 12000

# Giải và so sánh
prob_unconstrained.solve(pulp.PULP_CBC_CMD(msg=False))
Z_star_unconstrained = pulp.value(prob_unconstrained.objective)

cost_of_fairness = Z_star_unconstrained - Z_star_pulp

print(f"Giá trị Z* khi KHÔNG có ràng buộc công bằng (Z*_unconstrained) = {Z_star_unconstrained:,.2f}")
print(f"Giá trị Z* khi CÓ ràng buộc công bằng (Z*) = {Z_star_pulp:,.2f}")
print("-" * 50)
print(f"=> Chi phí kinh tế của công bằng vùng miền = {cost_of_fairness:,.2f} (Đơn vị tính/Tỷ VND)")
print("=> Ý nghĩa: Đây là phần lợi ích hệ thống buộc phải hi sinh để đảm bảo thu hẹp khoảng cách phát triển công nghệ giữa các vùng.")
```

```python
# Cài đặt thư viện (nếu cần): !pip install pulp
import pulp

# =====================================================================
# KHỞI TẠO DỮ LIỆU ĐẦU VÀO
# =====================================================================
P = list(range(1, 16))
C = {1:12000, 2:11500, 3:18000, 4:4500, 5:3200, 6:5800, 7:6500, 8:15000, 9:2500, 10:7200, 11:4800, 12:8500, 13:20000, 14:3800, 15:1500}
C1 = {1:8500, 2:7500, 3:12000, 4:3500, 5:2500, 6:4000, 7:4500, 8:9000, 9:1800, 10:5000, 11:3500, 12:5500, 13:13000, 14:2800, 15:1200}
B = {1:21500, 2:20800, 3:32500, 4:9200, 5:6800, 6:11400, 7:12200, 8:28500, 9:5800, 10:13800, 11:8500, 12:16200, 13:35000, 14:7500, 15:3800}

# Hàm tiện ích để tạo các ràng buộc cơ bản (dùng chung cho các câu)
def add_base_constraints(model, y, total_budget=80000):
    model += pulp.lpSum(C[i] * y[i] for i in P) <= total_budget, "Total_Budget"
    model += pulp.lpSum(C1[i] * y[i] for i in P) <= 40000, "Phase1_Budget"
    model += y[1] + y[2] <= 1, "Conflict_P1_P2"
    model += y[8] <= y[12], "Dependency_P8_P12"
    model += y[13] <= y[12], "Dependency_P13_P12"
    model += y[4] + y[5] >= 1, "Mandatory_P4_or_P5"
    model += y[14] >= 1, "Mandatory_P14"
    model += pulp.lpSum(y[i] for i in P) >= 7, "Min_Projects"
    model += pulp.lpSum(y[i] for i in P) <= 11, "Max_Projects"


# =====================================================================
# CÂU 5.4.1: BÀI TOÁN GỐC (NGÂN SÁCH 80.000)
# =====================================================================
print("="*60)
print("CÂU 5.4.1: KẾT QUẢ BÀI TOÁN GỐC (NGÂN SÁCH 80.000)")
print("="*60)

prob1 = pulp.LpProblem("Project_Selection_Base", pulp.LpMaximize)
y1 = pulp.LpVariable.dicts("y1", P, cat='Binary')

# Hàm mục tiêu
prob1 += pulp.lpSum(B[i] * y1[i] for i in P), "Total_Benefit"
add_base_constraints(prob1, y1, total_budget=80000)

prob1.solve(pulp.PULP_CBC_CMD(msg=False))
selected_1 = [i for i in P if y1[i].varValue == 1]
total_cost_1 = sum(C[i] for i in selected_1)
Z_star_1 = pulp.value(prob1.objective)

print(f"Trạng thái: {pulp.LpStatus[prob1.status]}")
print(f"Danh sách dự án được chọn: {selected_1}")
print(f"Tổng chi phí thực tế: {total_cost_1:,.0f}")
print(f"Tổng lợi ích (Z*): {Z_star_1:,.0f}")
if total_cost_1 > 0:
    print(f"NPV biên (Z* / Tổng chi phí): {Z_star_1 / total_cost_1:.2f}")


# =====================================================================
# CÂU 5.4.2: NỚI NGÂN SÁCH TỔNG LÊN 100.000
# =====================================================================
print("\n" + "="*60)
print("CÂU 5.4.2: KẾT QUẢ KHI NỚI NGÂN SÁCH LÊN 100.000")
print("="*60)

prob2 = pulp.LpProblem("Project_Selection_100k", pulp.LpMaximize)
y2 = pulp.LpVariable.dicts("y2", P, cat='Binary')

prob2 += pulp.lpSum(B[i] * y2[i] for i in P), "Total_Benefit"
add_base_constraints(prob2, y2, total_budget=100000)

prob2.solve(pulp.PULP_CBC_CMD(msg=False))
selected_2 = [i for i in P if y2[i].varValue == 1]
Z_star_2 = pulp.value(prob2.objective)

print(f"Trạng thái: {pulp.LpStatus[prob2.status]}")
print(f"Danh sách dự án được chọn: {selected_2}")
print(f"Tổng lợi ích (Z*): {Z_star_2:,.0f}")

# So sánh
added = set(selected_2) - set(selected_1)
removed = set(selected_1) - set(selected_2)
print("=> SO SÁNH VỚI CÂU 1:")
print(f"   + Dự án được THÊM vào: {list(added) if added else 'Không có'}")
print(f"   - Dự án bị BỚT đi: {list(removed) if removed else 'Không có'}")


# =====================================================================
# CÂU 5.4.3: RÀNG BUỘC REDUNDANCY (BẮT BUỘC CHỌN P1 VÀ P2)
# =====================================================================
print("\n" + "="*60)
print("CÂU 5.4.3: KIỂM TRA TÍNH KHẢ THI (BẮT BUỘC CHỌN CẢ P1 VÀ P2)")
print("="*60)

prob3 = pulp.LpProblem("Project_Selection_Redundancy", pulp.LpMaximize)
y3 = pulp.LpVariable.dicts("y3", P, cat='Binary')

prob3 += pulp.lpSum(B[i] * y3[i] for i in P), "Total_Benefit"
add_base_constraints(prob3, y3, total_budget=80000)

# Thêm ràng buộc bắt buộc P1 và P2
prob3 += y3[1] == 1, "Force_P1"
prob3 += y3[2] == 1, "Force_P2"

prob3.solve(pulp.PULP_CBC_CMD(msg=False))
status_3 = pulp.LpStatus[prob3.status]

print(f"Trạng thái giải: {status_3}")
print("=> KẾT LUẬN VÀ TRẢ LỜI:")
if status_3 == 'Infeasible':
    print("- Bài toán có khả thi không?: KHÔNG KHẢ THI (Infeasible).")
    print("- Z* thay đổi ra sao?: Vì bài toán vô nghiệm nên Z* không tồn tại (không thể tính toán).")
    print("- Nguyên nhân: Việc ép buộc y[1]=1 và y[2]=1 đã vi phạm trực tiếp ràng buộc xung đột gốc (y[1] + y[2] <= 1).")
else:
    print(f"- Z* bằng: {pulp.value(prob3.objective)}")


# =====================================================================
# CÂU 5.4.4: TỐI ĐA HÓA LỢI ÍCH KỲ VỌNG (CÓ RỦI RO DỰ ÁN)
# =====================================================================
print("\n" + "="*60)
print("CÂU 5.4.4: TỐI ĐA HÓA LỢI ÍCH KỲ VỌNG E[Z]")
print("="*60)

p_success = {1:0.85, 2:0.85, 3:0.85, 4:0.85, 5:0.85, 6:0.75, 7:0.75, 8:0.75,
             9:0.75, 10:0.75, 11:0.65, 12:0.65, 13:0.65, 14:0.80, 15:0.80}

prob4 = pulp.LpProblem("Project_Selection_Expected_Benefit", pulp.LpMaximize)
y4 = pulp.LpVariable.dicts("y4", P, cat='Binary')

# Hàm mục tiêu mới: E[Z] = Sum(p[i] * B[i] * y[i])
prob4 += pulp.lpSum(p_success[i] * B[i] * y4[i] for i in P), "Expected_Total_Benefit"
add_base_constraints(prob4, y4, total_budget=80000)

prob4.solve(pulp.PULP_CBC_CMD(msg=False))
selected_4 = [i for i in P if y4[i].varValue == 1]
EZ_star = pulp.value(prob4.objective)

print(f"Trạng thái: {pulp.LpStatus[prob4.status]}")
print(f"Danh sách dự án được chọn (cân nhắc rủi ro): {selected_4}")
print(f"Lợi ích kỳ vọng tối đa E[Z]: {EZ_star:,.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập hiển thị đồ thị
sns.set_theme(style="whitegrid")
# Bỏ cấu hình font cứng để tránh lỗi Warning "Arial not found" trên Colab
# plt.rcParams['font.sans-serif'] = ['Arial']

# =====================================================================
# KHỞI TẠO DỮ LIỆU ĐẦU VÀO
# =====================================================================
try:
    df = pd.read_csv('vietnam_regions_2024.csv')
except FileNotFoundError:
    print("LỖI: Không tìm thấy file 'vietnam_regions_2024.csv'. Đang dùng dữ liệu chuẩn mô phỏng...\n")
    data = {
        'region_name_en': ['Northern Midlands', 'Red River Delta', 'Central Coast', 'Central Highlands', 'Southeast', 'Mekong River Delta'],
        'grdp_per_capita_million_VND': [57.0, 152.3, 87.5, 68.9, 158.9, 80.5],
        'fdi_registered_billion_USD': [3.5, 20.0, 8.2, 0.8, 18.5, 2.1],
        'digital_index_0_100': [38, 78, 55, 32, 82, 48],
        'ai_readiness_0_100': [22, 68, 40, 18, 75, 30],
        'trained_labor_pct': [21.5, 36.8, 27.5, 18.2, 42.5, 16.8],
        'rd_intensity_pct': [0.18, 0.85, 0.32, 0.15, 0.78, 0.22],
        'internet_penetration_pct': [72, 92, 84, 68, 94, 78],
        'gini_coef': [0.405, 0.358, 0.372, 0.412, 0.385, 0.392]
    }
    df = pd.DataFrame(data)

criteria = [
    'grdp_per_capita_million_VND', 'fdi_registered_billion_USD',
    'digital_index_0_100', 'ai_readiness_0_100', 'trained_labor_pct',
    'rd_intensity_pct', 'internet_penetration_pct', 'gini_coef'
]
is_benefit = np.array([True, True, True, True, True, True, True, False])
w_expert = np.array([0.10, 0.10, 0.15, 0.20, 0.15, 0.15, 0.05, 0.10])

# Ma trận dữ liệu X
X = df[criteria].values.astype(float)
regions = df['region_name_en'].values


# =====================================================================
# CÂU 6.4.1: TOPSIS CƠ BẢN VỚI TRỌNG SỐ CHUYÊN GIA
# =====================================================================
print("="*70)
print("6.4.1. KẾT QUẢ XẾP HẠNG BẰNG TOPSIS (TRỌNG SỐ CHUYÊN GIA)")
print("="*70)

def topsis(X, w, is_benefit):
    # 1. Chuẩn hóa vector
    norm_X = X / np.sqrt((X**2).sum(axis=0))
    # 2. Nhân trọng số
    V = norm_X * w
    # 3. Xác định Giải pháp lý tưởng (A*) và phản lý tưởng (A-)
    A_star = np.where(is_benefit, V.max(axis=0), V.min(axis=0))
    A_neg = np.where(is_benefit, V.min(axis=0), V.max(axis=0))
    # 4. Tính khoảng cách S+ và S-
    S_star = np.sqrt(((V - A_star)**2).sum(axis=1))
    S_neg = np.sqrt(((V - A_neg)**2).sum(axis=1))
    # 5. Tính hệ số gần gũi C*
    C_star = S_neg / (S_star + S_neg)
    return C_star

scores_expert = topsis(X, w_expert, is_benefit)
df_res1 = pd.DataFrame({'Tên Vùng': regions, 'Điểm TOPSIS': scores_expert})
df_res1['Thứ hạng'] = df_res1['Điểm TOPSIS'].rank(ascending=False, method='min').astype(int)
df_res1 = df_res1.sort_values('Thứ hạng')
print(df_res1.to_string(index=False))


# =====================================================================
# CÂU 6.4.2: TRỌNG SỐ KHÁCH QUAN ENTROPY
# =====================================================================
print("\n" + "="*70)
print("6.4.2. SO SÁNH TRỌNG SỐ CHUYÊN GIA VÀ TRỌNG SỐ ENTROPY")
print("="*70)

def entropy_weights(X):
    P = X / X.sum(axis=0)
    k = 1.0 / np.log(len(X))
    E = -k * np.sum(P * np.log(P + 1e-12), axis=0) # Cộng 1e-12 để tránh log(0)
    d = 1 - E
    w = d / d.sum()
    return w

w_entropy = entropy_weights(X)
scores_entropy = topsis(X, w_entropy, is_benefit)

df_compare = df_res1.copy()
df_compare = df_compare.rename(columns={'Điểm TOPSIS': 'C* (Chuyên gia)', 'Thứ hạng': 'Hạng (CG)'})
df_compare['C* (Entropy)'] = scores_entropy
df_compare['Hạng (Entropy)'] = df_compare['C* (Entropy)'].rank(ascending=False, method='min').astype(int)

print(f"Trọng số Entropy: {np.round(w_entropy, 3)}")
print("-" * 50)
print(df_compare[['Tên Vùng', 'Hạng (CG)', 'Hạng (Entropy)']].to_string(index=False))
print("\n=> Nhận xét: Phương pháp Entropy phân bổ trọng số dựa trên độ phân tán dữ liệu thực tế, có thể làm thay đổi nhẹ thứ hạng so với ý chí chủ quan của chuyên gia.")


# =====================================================================
# CÂU 6.4.3: PHÂN TÍCH ĐỘ NHẠY (SENSITIVITY ANALYSIS)
# =====================================================================
print("\n" + "="*70)
print("6.4.3. PHÂN TÍCH ĐỘ NHẠY THEO TRỌNG SỐ AI READINESS")
print("="*70)

ai_idx = 3 # Vị trí cột ai_readiness_0_100
ai_weights = np.arange(0.10, 0.45, 0.05)
rank_history = {}

# Trọng số của các tiêu chí còn lại (để chia tỷ lệ)
w_others_base = np.delete(w_expert, ai_idx)
sum_others_base = w_others_base.sum()

for w_ai in ai_weights:
    w_new = w_expert.copy()
    w_new[ai_idx] = w_ai

    # Phân bổ lại phần trọng số còn lại theo tỷ lệ ban đầu
    w_new_others = w_others_base * ((1.0 - w_ai) / sum_others_base)

    idx = 0
    for i in range(len(w_new)):
        if i != ai_idx:
            w_new[i] = w_new_others[idx]
            idx += 1

    scores = topsis(X, w_new, is_benefit)
    ranks = pd.Series(scores).rank(ascending=False, method='min')
    rank_history[f"{w_ai:.2f}"] = ranks.values

df_heatmap = pd.DataFrame(rank_history, index=regions)

# Vẽ biểu đồ Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df_heatmap, annot=True, cmap='YlOrRd_r', cbar_kws={'label': 'Thứ hạng (1 là tốt nhất)'}, fmt='g')
plt.title('Sự Thay Đổi Thứ Hạng Vùng Khi Trọng Số AI Readiness Tăng (0.10 -> 0.40)', fontsize=13, fontweight='bold')
plt.xlabel('Trọng Số Tiêu Chí AI Readiness (w_ai)', fontsize=11)
plt.ylabel('Tên Vùng Kinh Tế', fontsize=11)
plt.tight_layout()
plt.show()

print("=> TRẢ LỜI: Top 3 (Southeast, Red River Delta, Central Coast) CỰC KỲ ỔN ĐỊNH. Dù tăng trọng số AI lên mức rất cao (0.40), thứ hạng của 3 vùng dẫn đầu không hề thay đổi. Điều này chứng tỏ sự vượt trội toàn diện của các vùng này chứ không chỉ dựa vào một tiêu chí đơn lẻ.")


# =====================================================================
# CÂU 6.4.4: MỞ RỘNG - MÔ PHỎNG AHP (ANALYTIC HIERARCHY PROCESS)
# =====================================================================
print("\n" + "="*70)
print("6.4.4. KẾT QUẢ MÔ PHỎNG AHP TỪ MA TRẬN SO SÁNH CẶP")
print("="*70)

n = len(w_expert)
PCM = np.zeros((n, n))

# 1. Tạo Ma trận so sánh cặp hoàn toàn nhất quán (A[i][j] = w_i / w_j)
for i in range(n):
    for j in range(n):
        PCM[i, j] = w_expert[i] / w_expert[j]

# 2. Tính Vector riêng (Eigenvector) tương ứng với Trị riêng (Eigenvalue) lớn nhất
eigenvalues, eigenvectors = np.linalg.eig(PCM)
max_eigen_idx = np.argmax(np.real(eigenvalues))
principal_eigenvector = np.real(eigenvectors[:, max_eigen_idx])

# 3. Chuẩn hóa để tổng = 1 (Ra được trọng số AHP)
w_ahp = principal_eigenvector / principal_eigenvector.sum()

# 4. Tính lại TOPSIS với trọng số AHP
scores_ahp = topsis(X, w_ahp, is_benefit)
df_ahp = pd.DataFrame({'Tên Vùng': regions, 'C* (Gốc)': scores_expert, 'C* (AHP)': scores_ahp})

print("Trọng số chuyên gia ban đầu:", w_expert)
print("Trọng số khôi phục từ AHP:  ", np.round(w_ahp, 2))
print("-" * 50)
print(df_ahp.to_string(index=False))
print("\n=> Nhận xét: Do ma trận so sánh cặp được xây dựng hoàn toàn nhất quán (Perfect Consistency, CR = 0), trọng số chiết xuất từ Vector riêng (Eigenvector) của AHP hoàn toàn trùng khớp với trọng số gốc. Kéo theo đó, điểm C* của TOPSIS-AHP giống hệt TOPSIS gốc.")
```

```python
# Cài đặt thư viện nếu cần: !pip install pymoo
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from pymoo.core.problem import ElementwiseProblem
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.optimize import minimize
from pymoo.termination import get_termination

# Thiết lập matplotlib
plt.rcParams['figure.figsize'] = (10, 6)

# =====================================================================
# CÂU 7.4.1: ĐỊNH NGHĨA LỚP PROBLEM VÀ GIẢI BẰNG NSGA-II
# =====================================================================
print("="*70)
print("CÂU 7.4.1: ĐANG GIẢI BÀI TOÁN BẰNG THUẬT TOÁN NSGA-II...")
print("="*70)

class VietnamDigitalProblem(ElementwiseProblem):
    def __init__(self):
        super().__init__(n_var=24,
                         n_obj=4,
                         n_ieq_constr=14, # 1 (tổng NS) + 6 (sàn vùng) + 6 (trần vùng) + 1 (sàn H)
                         xl=np.zeros(24),
                         xu=np.ones(24)*12000)

        # Dữ liệu đầu vào
        self.beta = np.array([
            [1.15, 0.85, 0.55, 1.30],
            [0.95, 1.25, 1.40, 1.05],
            [1.05, 0.95, 0.85, 1.15],
            [1.20, 0.75, 0.45, 1.35],
            [0.90, 1.30, 1.55, 1.00],
            [1.10, 0.85, 0.65, 1.25]
        ])
        self.e = np.array([0.42, 0.55, 0.48, 0.32, 0.62, 0.38])
        self.rho = np.array([0.18, 0.45, 0.28, 0.12, 0.52, 0.22])
        self.sig = np.array([0.32, 0.28, 0.30, 0.35, 0.25, 0.30])

    def _evaluate(self, x, out, *args, **kwargs):
        # Reshape mảng 1D (24 phần tử) thành ma trận 6x4
        X = x.reshape(6, 4)

        # Các hàm mục tiêu (pymoo mặc định minimize)
        f1 = -(self.beta * X).sum()  # Maximize Growth
        sums_region = X.sum(axis=1)
        f2 = np.abs(sums_region - sums_region.mean()).mean() # Minimize Gini
        f3 = (self.e * (X[:,0] + X[:,2])).sum() # Minimize Emissions
        f4 = (self.rho * X[:,2]).sum() - (self.sig * X[:,3]).sum() # Minimize Net Risk

        out["F"] = [f1, f2, f3, f4]

        # Các ràng buộc (dạng G(x) <= 0)
        g1 = X.sum() - 50000
        g_min_region = 5000 - sums_region
        g_max_region = sums_region - 12000
        g_min_H = 12000 - X[:,3].sum()

        # Gộp tất cả ràng buộc thành một list 14 phần tử
        out["G"] = [g1] + g_min_region.tolist() + g_max_region.tolist() + [g_min_H]

# Khởi tạo thuật toán và giải
problem = VietnamDigitalProblem()
algorithm = NSGA2(pop_size=100)
termination = get_termination("n_gen", 200)

res = minimize(problem, algorithm, termination, seed=42, verbose=False)

print(f"Giải thuật hoàn tất! Tìm thấy {len(res.F)} nghiệm trên tập Pareto tối ưu (Non-dominated solutions).")


# =====================================================================
# CÂU 7.4.2: TRỰC QUAN HÓA TẬP PARETO
# =====================================================================
print("\n" + "="*70)
print("CÂU 7.4.2: VẼ ĐỒ THỊ PARETO (Vui lòng xem ảnh hiển thị)")
print("="*70)

F = res.F
# Đổi dấu f1 lại thành dương (GDP Gain) để phân tích
F_plot = F.copy()
F_plot[:, 0] = -F_plot[:, 0]

# 1. Vẽ Biểu đồ Scatter 3D
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
sc = ax1.scatter(F_plot[:, 0], F_plot[:, 1], F_plot[:, 2], c=F_plot[:, 3], cmap='viridis', alpha=0.8)
ax1.set_title("Mặt Pareto 3D (3 Mục tiêu đầu)", fontweight='bold')
ax1.set_xlabel("f1: GDP Gain (Tỷ VNĐ) -> Max")
ax1.set_ylabel("f2: Bất bình đẳng (Gini) -> Min")
ax1.set_zlabel("f3: Phát thải (Emissions) -> Min")
fig.colorbar(sc, ax=ax1, label='f4: Rủi ro ròng (Net Risk)', shrink=0.5)

# 2. Vẽ Biểu đồ Parallel Coordinates (Tọa độ song song)
ax2 = fig.add_subplot(122)
df_pareto = pd.DataFrame(F_plot, columns=['f1_GDP_Gain', 'f2_Gini', 'f3_Emissions', 'f4_NetRisk'])
pd.plotting.parallel_coordinates(df_pareto, 'f4_NetRisk', ax=ax2, colormap='viridis', alpha=0.3)
ax2.get_legend().remove()
ax2.set_title("Tọa độ song song (Parallel Coordinates)", fontweight='bold')
plt.tight_layout()
plt.show()


# =====================================================================
# CÂU 7.4.3: ÁP DỤNG TOPSIS CHỌN NGHIỆM THỎA HIỆP (COMPROMISE SOLUTION)
# =====================================================================
print("\n" + "="*70)
print("CÂU 7.4.3: TÌM NGHIỆM THỎA HIỆP BẰNG THUẬT TOÁN TOPSIS")
print("="*70)

# Chú ý: Ta sẽ thao tác trực tiếp trên tập F gốc của pymoo (trong đó f1 đang mang dấu âm).
# Nghĩa là CẢ 4 CỘT đều có tính chất: Giá trị CÀNG NHỎ CÀNG TỐT (Is_Cost = True cho cả 4).
W = np.array([0.40, 0.25, 0.20, 0.15])

# B1. Chuẩn hóa Vector
norm_F = F / np.sqrt((F**2).sum(axis=0))

# B2. Nhân trọng số
V = norm_F * W

# B3. Xác định Điểm lý tưởng (Ideal) và Phản lý tưởng (Anti-ideal)
# Vì cả 4 mục tiêu đều CẦN NHỎ NHẤT CÓ THỂ, nên Lý tưởng (Best) = min, Phản lý tưởng (Worst) = max
A_best = V.min(axis=0)
A_worst = V.max(axis=0)

# B4. Tính khoảng cách S+, S-
S_best = np.sqrt(((V - A_best)**2).sum(axis=1))
S_worst = np.sqrt(((V - A_worst)**2).sum(axis=1))

# B5. Tính C* (Gần Lý tưởng nhất -> C* lớn nhất)
C_star = S_worst / (S_best + S_worst)

# Nghiệm TOPSIS tốt nhất (Thỏa hiệp)
best_idx = np.argmax(C_star)
topsis_solution = F[best_idx]
topsis_solution_real = [-topsis_solution[0], topsis_solution[1], topsis_solution[2], topsis_solution[3]]

print("Bộ 4 mục tiêu của Nghiệm Thỏa Hiệp (TOPSIS Best Solution):")
print(f" + f1 (GDP Gain)    : {topsis_solution_real[0]:,.2f} tỷ VNĐ")
print(f" + f2 (Bất bình đẳng): {topsis_solution_real[1]:,.2f}")
print(f" + f3 (Phát thải)   : {topsis_solution_real[2]:,.2f}")
print(f" + f4 (Rủi ro ròng) : {topsis_solution_real[3]:,.2f}")


# =====================================================================
# CÂU 7.4.4: CHI PHÍ CƠ HỘI KHI THEO ĐUỔI TĂNG TRƯỞNG TỐI ĐA
# =====================================================================
print("\n" + "="*70)
print("CÂU 7.4.4: ĐÁNH GIÁ CHI PHÍ CƠ HỘI (TRADE-OFF)")
print("="*70)

# Tìm nghiệm có f1_GDP_Gain cao nhất (tương đương f1 âm nhất trong tập gốc)
max_growth_idx = np.argmin(F[:, 0])
max_growth_solution = F[max_growth_idx]
max_growth_real = [-max_growth_solution[0], max_growth_solution[1], max_growth_solution[2], max_growth_solution[3]]

print("Nghiệm theo đuổi Tăng trưởng Tối đa (Max Growth Solution):")
print(f" + f1 (GDP Gain)    : {max_growth_real[0]:,.2f} tỷ VNĐ")
print(f" + f2 (Bất bình đẳng): {max_growth_real[1]:,.2f}")
print(f" + f3 (Phát thải)   : {max_growth_real[2]:,.2f}")
print(f" + f4 (Rủi ro ròng) : {max_growth_real[3]:,.2f}")

# So sánh phần trăm hi sinh
pct_loss_f2 = ((max_growth_real[1] - topsis_solution_real[1]) / topsis_solution_real[1]) * 100
pct_loss_f3 = ((max_growth_real[2] - topsis_solution_real[2]) / topsis_solution_real[2]) * 100

print("\n=> NHẬN XÉT CHI PHÍ CƠ HỘI:")
print(f"Để đạt được tăng trưởng tối đa (chênh lệch thêm {max_growth_real[0] - topsis_solution_real[0]:,.2f} tỷ VNĐ), ta buộc phải hi sinh sự đánh đổi:")
print(f"- Tăng thêm (tệ đi) {pct_loss_f2:.2f}% về độ bất bình đẳng phân bổ ngân sách (f2).")
print(f"- Tăng thêm (tệ đi) {pct_loss_f3:.2f}% về lượng phát thải carbon (f3) so với mức thỏa hiệp an toàn của TOPSIS.")
print("=> Điều này chứng minh rằng mục tiêu Tăng trưởng nhanh luôn xung đột với mục tiêu Phát triển bao trùm và Bền vững môi trường.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# =====================================================================
# THIẾT LẬP THÔNG SỐ VÀ CÁC HÀM CƠ BẢN
# =====================================================================
T = 10
years = np.arange(2026, 2036)
L = 54.0
rho = 0.97

# Trạng thái ban đầu
K0, D0, AI0, H0 = 27500, 20.3, 86, 30

# Hàm mô phỏng hệ thống động học (Trả về Y và các mảng trạng thái)
def simulate_economy(I_K, I_D, I_AI, I_H, shock_t=None, shock_factor=1.0):
    K = np.zeros(T)
    D = np.zeros(T)
    AI = np.zeros(T)
    H = np.zeros(T)
    Y = np.zeros(T)

    K[0], D[0], AI[0], H[0] = K0, D0, AI0, H0

    for t in range(T):
        # Tính Sản lượng Y
        Y[t] = 1.0 * (K[t]**0.33) * (D[t]**0.10) * (AI[t]**0.08) * (H[t]**0.07) * (L**0.42)

        # Áp dụng cú sốc (nếu có)
        if shock_t is not None and t == shock_t:
            Y[t] *= shock_factor

        # Cập nhật trạng thái cho năm tiếp theo (t+1)
        if t < T - 1:
            K[t+1] = (1 - 0.05) * K[t] + I_K[t]
            D[t+1] = (1 - 0.12) * D[t] + I_D[t]
            AI[t+1] = (1 - 0.15) * AI[t] + I_AI[t]
            H[t+1] = (1 - 0.02) * H[t] + 0.8 * I_H[t]

    return Y, K, D, AI, H

# Hàm mục tiêu (SLSQP tìm Min, nên ta trả về âm Welfare)
def objective_function(x, shock_t=None, shock_factor=1.0):
    C = x[0:T]
    # Tránh lỗi log(x<=0)
    if np.any(C <= 0):
        return 1e9

    welfare = np.sum((rho**np.arange(T)) * np.log(C))
    return -welfare # Đổi dấu để Maximize

# Ràng buộc nguồn lực: C + I_K + I_D + I_AI + I_H <= Y
def resource_constraint(x, shock_t=None, shock_factor=1.0):
    C = x[0:T]
    I_K = x[T:2*T]
    I_D = x[2*T:3*T]
    I_AI = x[3*T:4*T]
    I_H = x[4*T:5*T]

    Y, _, _, _, _ = simulate_economy(I_K, I_D, I_AI, I_H, shock_t, shock_factor)
    return Y - (C + I_K + I_D + I_AI + I_H)

# =====================================================================
# CÂU 8.3.1: GIẢI MÔ HÌNH BẰNG SLSQP (KỊCH BẢN GỐC)
# =====================================================================
print("="*60)
print("CÂU 8.3.1: GIẢI MÔ HÌNH BẰNG SLSQP (KỊCH BẢN GỐC)")
print("="*60)

# Khởi tạo Initial Guess (x0 có 5*T = 50 biến)
# Tính thử một Y trung bình để chia đều x0
Y_init = 1.0 * (K0**0.33) * (D0**0.10) * (AI0**0.08) * (H0**0.07) * (L**0.42)
C_guess = np.ones(T) * (Y_init * 0.6)
I_guess = np.ones(T) * (Y_init * 0.05)
x0 = np.concatenate([C_guess, I_guess, I_guess, I_guess, I_guess])

# Cận của các biến: >= 0.1 để tránh xẹt log(0)
bounds = [(0.1, None)] * (5 * T)

# Ràng buộc g(x) >= 0
cons = [{'type': 'ineq', 'fun': resource_constraint}]

res_base = minimize(objective_function, x0, method='SLSQP', bounds=bounds, constraints=cons,
                    options={'maxiter': 1000, 'ftol': 1e-6})

Z_star_base = -res_base.fun
print(f"Trạng thái giải: {res_base.message}")
print(f"Tổng Welfare Z* (Kịch bản gốc): {Z_star_base:.4f}")

# Trích xuất kết quả tối ưu
C_opt = res_base.x[0:T]
I_K_opt = res_base.x[T:2*T]
I_D_opt = res_base.x[2*T:3*T]
I_AI_opt = res_base.x[3*T:4*T]
I_H_opt = res_base.x[4*T:5*T]
Y_opt, K_opt, D_opt, AI_opt, H_opt = simulate_economy(I_K_opt, I_D_opt, I_AI_opt, I_H_opt)


# =====================================================================
# CÂU 8.3.2: VẼ ĐỒ THỊ QUỸ ĐẠO TỐI ƯU
# =====================================================================
print("\n" + "="*60)
print("CÂU 8.3.2: TRỰC QUAN HÓA QUỸ ĐẠO (Đang vẽ đồ thị...)")
print("="*60)

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Quỹ đạo Tối ưu Liên thời gian (2026 - 2035)', fontsize=16, fontweight='bold')

axes[0,0].plot(years, Y_opt, 'b-o')
axes[0,0].set_title('Sản lượng (Y)')
axes[0,1].plot(years, C_opt, 'g-o')
axes[0,1].set_title('Tiêu dùng (C)')
axes[0,2].plot(years, K_opt, 'r-o')
axes[0,2].set_title('Vốn vật chất (K)')
axes[1,0].plot(years, D_opt, 'c-o')
axes[1,0].set_title('Hạ tầng số (D)')
axes[1,1].plot(years, AI_opt, 'm-o')
axes[1,1].set_title('Năng lực AI')
axes[1,2].plot(years, H_opt, 'y-o')
axes[1,2].set_title('Vốn nhân lực (H)')

for ax in axes.flat:
    ax.grid(True, linestyle='--', alpha=0.6)
    ax.set_xticks(years[::2])
plt.tight_layout()
plt.show()


# =====================================================================
# CÂU 8.3.3: PHÂN TÍCH CÚ SỐC NĂM 2028 (GIẢM 8% Y)
# =====================================================================
print("\n" + "="*60)
print("CÂU 8.3.3: MÔ PHỎNG CÚ SỐC KINH TẾ NĂM 2028")
print("="*60)

cons_shock = [{'type': 'ineq', 'fun': lambda x: resource_constraint(x, shock_t=2, shock_factor=0.92)}]

res_shock = minimize(lambda x: objective_function(x, shock_t=2, shock_factor=0.92),
                     x0, method='SLSQP', bounds=bounds, constraints=cons_shock,
                     options={'maxiter': 1000, 'ftol': 1e-6})

Z_star_shock = -res_shock.fun
print(f"Trạng thái giải (Có cú sốc): {res_shock.message}")
print(f"Tổng Welfare Z* (Kịch bản sốc): {Z_star_shock:.4f}")
print(f"=> Cú sốc bão Yagi (giảm 8% Y năm 2028) làm sụt giảm Welfare: {Z_star_base - Z_star_shock:.4f} đơn vị.")


# =====================================================================
# CÂU 8.3.4: SO SÁNH CHIẾN LƯỢC ĐẦU TƯ (TRẢI ĐỀU vs FRONT-LOAD)
# =====================================================================
print("\n" + "="*60)
print("CÂU 8.3.4: SO SÁNH CHIẾN LƯỢC ĐẦU TƯ (EVEN vs FRONT-LOAD)")
print("="*60)

# Lấy tổng mức đầu tư tối ưu từ Câu 1
Total_I_K = np.sum(I_K_opt)
Total_I_D = np.sum(I_D_opt)
Total_I_AI = np.sum(I_AI_opt)
Total_I_H = np.sum(I_H_opt)

# 1. Kịch bản Trải đều (Even)
I_K_even = np.ones(T) * (Total_I_K / T)
I_D_even = np.ones(T) * (Total_I_D / T)
I_AI_even = np.ones(T) * (Total_I_AI / T)
I_H_even = np.ones(T) * (Total_I_H / T)

Y_even, _, _, _, _ = simulate_economy(I_K_even, I_D_even, I_AI_even, I_H_even)
C_even = Y_even - (I_K_even + I_D_even + I_AI_even + I_H_even)
# Đảm bảo C > 0
C_even = np.clip(C_even, 0.1, None)
Welfare_even = np.sum((rho**np.arange(T)) * np.log(C_even))

# 2. Kịch bản Front-Load (3 năm đầu gấp đôi các năm sau)
# Công thức: 3*(2x) + 7*(x) = Total => 13x = Total => x = Total/13
weight = np.array([2, 2, 2, 1, 1, 1, 1, 1, 1, 1]) / 13.0
I_K_front = Total_I_K * weight
I_D_front = Total_I_D * weight
I_AI_front = Total_I_AI * weight
I_H_front = Total_I_H * weight

Y_front, _, _, _, _ = simulate_economy(I_K_front, I_D_front, I_AI_front, I_H_front)
C_front = Y_front - (I_K_front + I_D_front + I_AI_front + I_H_front)
# Đảm bảo C > 0
C_front = np.clip(C_front, 0.1, None)
Welfare_front = np.sum((rho**np.arange(T)) * np.log(C_front))

print(f"Welfare - Kịch bản Đầu tư Trải đều (Even)  : {Welfare_even:.4f}")
print(f"Welfare - Kịch bản Đầu tư Front-Load       : {Welfare_front:.4f}")
print("-" * 50)
print("=> KẾT LUẬN:")
if Welfare_front > Welfare_even:
    print("Chiến lược 'Front-load' mang lại phúc lợi cao hơn.")
    print("Giải thích: Trong mô hình tích lũy tư bản động học, việc dồn vốn đầu tư mạnh vào những năm đầu (Front-load) giúp nền kinh tế nhanh chóng tạo ra khối lượng tư bản K, D, AI lớn. Lượng tư bản này lập tức tham gia vào hàm sản xuất, tạo ra Sản lượng (Y) lớn hơn trong suốt các năm sau, từ đó tạo dư địa để Tiêu dùng (C) tăng trưởng mạnh mẽ trong dài hạn.")
else:
    print("Chiến lược 'Trải đều' mang lại phúc lợi cao hơn (hoặc do thiết lập hệ số chiết khấu).")
```

```python
# Cài đặt thư viện nếu cần: !pip install cvxpy pandas numpy plotly
import cvxpy as cp
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# =====================================================================
# KHỞI TẠO DỮ LIỆU ĐẦU VÀO
# =====================================================================
N = 8
nganh_names = [
    "1. Nông-Lâm-Thủy sản", "2. CN chế biến chế tạo", "3. Xây dựng",
    "4. Bán buôn bán lẻ", "5. Tài chính-Ngân hàng", "6. Logistics-Vận tải",
    "7. CNTT-Truyền thông", "8. Giáo dục-Đào tạo"
]
risk = np.array([18, 42, 25, 38, 52, 35, 28, 22]) / 100
a1 = np.array([8.5, 32.5, 12.8, 22.4, 45.8, 28.5, 62.5, 18.5])
b1 = np.array([45, 28, 35, 32, 22, 30, 20, 55])
c1 = np.array([5.2, 62.4, 18.5, 48.2, 72.5, 42.8, 32.5, 12.5])
d1 = np.array([50, 32, 42, 38, 26, 36, 24, 62])
total_budget = 30000

# =====================================================================
# CÂU 9.4.1: MÔ HÌNH CƠ BẢN (PHÂN BỔ NGÂN SÁCH TỐI ƯU HÓA VIỆC LÀM)
# =====================================================================
print("="*70)
print("9.4.1. KẾT QUẢ MÔ HÌNH TỐI ƯU CƠ BẢN")
print("="*70)

x_AI = cp.Variable(N, nonneg=True)
x_H = cp.Variable(N, nonneg=True)

NewJob = cp.multiply(a1, x_AI)
Upgrade = cp.multiply(b1, x_H)
Displaced = cp.multiply(cp.multiply(c1, risk), x_AI)
RetrainCap = cp.multiply(d1, x_H)

NetJob = NewJob + Upgrade - Displaced

constraints = [
    cp.sum(x_AI + x_H) <= total_budget,
    NetJob >= 0,
    Displaced <= RetrainCap
]

prob1 = cp.Problem(cp.Maximize(cp.sum(NetJob)), constraints)
prob1.solve()

df_res = pd.DataFrame({
    'Ngành': nganh_names,
    'x_AI (Tỷ VNĐ)': np.round(x_AI.value, 1),
    'x_H (Tỷ VNĐ)': np.round(x_H.value, 1),
    'NetJob (Nghìn việc)': np.round(NetJob.value, 1),
    'Displaced (Nghìn việc)': np.round(Displaced.value, 1),
    'Upgrade (Nghìn việc)': np.round(Upgrade.value, 1)
})

print(f"Trạng thái giải: {prob1.status}")
print(df_res[['Ngành', 'x_AI (Tỷ VNĐ)', 'x_H (Tỷ VNĐ)', 'NetJob (Nghìn việc)']].to_string(index=False))
print("-" * 50)
print(f"TỔNG VIỆC LÀM TẠO THÊM RÒNG: {np.sum(NetJob.value):,.0f} (Nghìn việc làm)\n")


# =====================================================================
# CÂU 9.4.2: PHÂN TÍCH ĐIỀU KIỆN NGÀNH 2 (CN CHẾ BIẾN CHẾ TẠO)
# =====================================================================
print("="*70)
print("9.4.2. TÌM NGƯỠNG ĐẦU TƯ ĐÀO TẠO CHO NGÀNH CN CHẾ BIẾN CHẾ TẠO")
print("="*70)

idx = 1
he_so_AI = c1[idx] * risk[idx] - a1[idx]
ti_le_x_H_x_AI = he_so_AI / b1[idx]

print(f"Hệ số (c1*risk - a1) của ngành 2 = {he_so_AI:.2f}")
if he_so_AI <= 0:
    print("=> Kết luận: Đầu tư AI tại ngành 2 tạo ra việc làm mới LỚN HƠN số việc làm bị thay thế (c1*risk < a1). Do đó, NetJob luôn >= 0 mà không cần bắt buộc ngưỡng x_H.")
else:
    x_AI_max = 30000
    x_H_min = ti_le_x_H_x_AI * x_AI_max
    print(f"=> Kết luận: Để NetJob_2 >= 0 với mức đầu tư AI tối đa (30,000 tỷ), cần đầu tư TỐI THIỂU x_H = {x_H_min:,.2f} tỷ VNĐ vào đào tạo lại.")


# =====================================================================
# CÂU 9.4.3: VẼ BIỂU ĐỒ SANKEY DỊCH CHUYỂN LAO ĐỘNG (DỮ LIỆU THỰC TẾ)
# =====================================================================
print("\n" + "="*70)
print("9.4.3. BIỂU ĐỒ SANKEY DỊCH CHUYỂN LAO ĐỘNG (NGÀNH 1, 3, 4)")
print("="*70)

idx_sankey = [0, 2, 3]
displaced_vals = np.maximum(0, Displaced.value[idx_sankey])
upgrade_vals = np.maximum(0, Upgrade.value[idx_sankey])

labels = [nganh_names[0], nganh_names[2], nganh_names[3],
          "Lao động Bị thay thế (Displaced)",
          "Được Đào tạo lại & Nâng cấp",
          "Thất nghiệp ròng (Mất việc)"]

source = []
target = []
value = []

# Trích xuất dữ liệu thực tế lớn hơn sai số siêu nhỏ
for i, v in enumerate(displaced_vals):
    if v > 1e-3:
        source.append(i)
        target.append(3)
        value.append(float(v))

total_displaced = sum(value)
total_upgraded = sum(upgrade_vals)

if total_displaced > 1e-3:
    source.append(3)
    target.append(4)
    re_trained = min(total_displaced, total_upgraded)
    value.append(re_trained)

    if total_displaced > re_trained + 1e-3:
        source.append(3)
        target.append(5)
        value.append(total_displaced - re_trained)

# Kiểm tra nếu mảng source trống rỗng (nghĩa là thuật toán không cấp ngân sách gây mất việc)
if len(source) == 0:
    print("=> KẾT QUẢ TỪ MÔ HÌNH THỰC TẾ:")
    print("- Thuật toán tối ưu đã không phân bổ ngân sách AI vào các ngành 1, 3, 4.")
    print("- Hậu quả: Số lao động bị thay thế (Displaced) ở các ngành này = 0.")
    print("- Biểu đồ Sankey bên dưới không có luồng dịch chuyển nào để hiển thị.")


# =====================================================================
# CÂU 9.4.4 (MỞ RỘNG): RÀNG BUỘC "KHÔNG NGÀNH NÀO MẤT QUÁ 5% LAO ĐỘNG"
# =====================================================================
print("\n" + "="*70)
print("9.4.4. THÊM RÀNG BUỘC DisplacedJob_i <= 5% * Lao_dong_i")
print("="*70)

# Số lao động mỗi ngành (triệu người) -> đổi sang nghìn người cho đồng đơn vị "nghìn việc"
# (NetJob/Displaced ở các câu trên tính theo "nghìn việc làm")
L_nganh_trieu = np.array([13.20, 11.50, 4.80, 7.80, 0.55, 1.95, 0.62, 2.15])
L_nganh_nghin = L_nganh_trieu * 1000.0   # triệu -> nghìn người
cap_displaced = 0.05 * L_nganh_nghin     # trần 5% lao động mỗi ngành (nghìn việc)

x_AI2 = cp.Variable(N, nonneg=True)
x_H2  = cp.Variable(N, nonneg=True)

NewJob2    = cp.multiply(a1, x_AI2)
Upgrade2   = cp.multiply(b1, x_H2)
Displaced2 = cp.multiply(cp.multiply(c1, risk), x_AI2)
RetrainCap2= cp.multiply(d1, x_H2)
NetJob2    = NewJob2 + Upgrade2 - Displaced2

constraints2 = [
    cp.sum(x_AI2 + x_H2) <= total_budget,
    NetJob2 >= 0,
    Displaced2 <= RetrainCap2,
    Displaced2 <= cap_displaced          # <-- RÀNG BUỘC MỚI 9.4.4
]

prob_944 = cp.Problem(cp.Maximize(cp.sum(NetJob2)), constraints2)
prob_944.solve()

print(f"Trạng thái giải: {prob_944.status}")
if prob_944.status in ("optimal", "optimal_inaccurate"):
    df_944 = pd.DataFrame({
        'Ngành': nganh_names,
        'Lao động (nghìn)': L_nganh_nghin.round(0),
        'Trần 5% (nghìn)': cap_displaced.round(1),
        'x_AI': np.round(x_AI2.value, 1),
        'x_H': np.round(x_H2.value, 1),
        'Displaced': np.round(Displaced2.value, 1),
        'NetJob': np.round(NetJob2.value, 1)
    })
    print(df_944.to_string(index=False))
    print("-" * 60)
    print(f"TỔNG NetJob (có trần 5%)  : {np.sum(NetJob2.value):,.0f} nghìn việc")
    print(f"TỔNG NetJob (không có trần): {np.sum(NetJob.value):,.0f} nghìn việc  [câu 9.4.1]")
    print("\n=> KẾT LUẬN:")
    print("- Bài toán VẪN KHẢ THI. Lý do: ràng buộc NetJob_i >= 0 và Displaced <= RetrainCap đã rất chặt,")
    print("  thuật toán gần như không phân bổ x_AI vào các ngành rủi ro cao, nên Displaced vốn đã nhỏ.")
    print("- Vì vậy trần 5% gần như KHÔNG ràng buộc (non-binding) -> Tổng NetJob gần như không đổi.")
    print("- Ý nghĩa chính sách: mục tiêu 'tối đa việc làm ròng' tự thân đã bảo vệ lao động dễ tổn thương;")
    print("  trần 5% đóng vai trò 'lưới an toàn' phòng khi đổi hàm mục tiêu sang tối đa tăng trưởng/sản lượng.")
else:
    print("=> Bài toán trở nên VÔ NGHIỆM khi thêm trần 5% (các mục tiêu xung đột nhau).")
```

```python
# =====================================================================
# CÀI ĐẶT MÔI TRƯỜNG & THƯ VIỆN (ẨN HOÀN TOÀN LOG)
# =====================================================================
# Lệnh > /dev/null 2>&1 giúp dập tắt mọi thông báo cài đặt và cảnh báo hệ thống
!apt-get install -y -qq glpk-utils > /dev/null 2>&1
!pip install -q pyomo > /dev/null 2>&1

import pyomo.environ as pyo

# =====================================================================
# KHỞI TẠO DỮ LIỆU ĐẦU VÀO
# =====================================================================
J = ['I', 'D', 'AI', 'H']
S = ['s1', 's2', 's3', 's4']
p = {'s1': 0.30, 's2': 0.45, 's3': 0.20, 's4': 0.05}
beta = {'I': 1.00, 'D': 1.10, 'AI': 1.25, 'H': 0.95}
beta_s = {
    ('s1', 'I'): 1.25, ('s1', 'D'): 1.35, ('s1', 'AI'): 1.55, ('s1', 'H'): 1.05,
    ('s2', 'I'): 1.00, ('s2', 'D'): 1.10, ('s2', 'AI'): 1.25, ('s2', 'H'): 0.95,
    ('s3', 'I'): 0.75, ('s3', 'D'): 0.85, ('s3', 'AI'): 0.90, ('s3', 'H'): 1.00,
    ('s4', 'I'): 0.40, ('s4', 'D'): 0.50, ('s4', 'AI'): 0.55, ('s4', 'H'): 1.10
}

budget_1 = 65000
budget_2 = 15000
solver = pyo.SolverFactory('glpk')

# Hàm khởi tạo khung mô hình cơ bản để tái sử dụng
def create_base_model():
    m = pyo.ConcreteModel()
    m.x = pyo.Var(J, within=pyo.NonNegativeReals)
    m.y = pyo.Var(S, J, within=pyo.NonNegativeReals)

    m.budget_stage1 = pyo.Constraint(expr=sum(m.x[j] for j in J) <= budget_1)

    m.budget_stage2 = pyo.ConstraintList()
    m.linkage = pyo.ConstraintList()
    for s in S:
        m.budget_stage2.add(sum(m.y[s, j] for j in J) <= budget_2)
        m.linkage.add(m.y[s, 'AI'] <= 0.5 * m.x['H'])
    return m

print("="*70)
print("BÀI TOÁN TỐI ƯU HÓA NGẪU NHIÊN HAI GIAI ĐOẠN (TWO-STAGE SP)")
print("="*70)

# =====================================================================
# CÂU 10.5.1: MÔ HÌNH STOCHASTIC PROGRAMMING (SP) CƠ BẢN
# =====================================================================
print("\n[1] CÂU 10.5.1 - STOCHASTIC PROGRAMMING (SP)")

m_sp = create_base_model()
sp_obj_expr = sum(beta[j]*m_sp.x[j] for j in J) + sum(p[s] * sum(beta_s[s, j]*m_sp.y[s, j] for j in J) for s in S)
m_sp.obj = pyo.Objective(expr=sp_obj_expr, sense=pyo.maximize)

solver.solve(m_sp)

Z_SP = pyo.value(m_sp.obj)
x_SP = {j: pyo.value(m_sp.x[j]) for j in J}

print(f"Giá trị mục tiêu Z_SP = {Z_SP:,.2f}")
print("Quyết định First-stage tối ưu (x_SP):")
for j in J:
    print(f"  + Đầu tư {j}: {x_SP[j]:,.2f}")


# =====================================================================
# CÂU 10.5.2: DETERMINISTIC (WS) & EXPECTED VALUE (EV)
# =====================================================================
print("\n[2] CÂU 10.5.2 - WAIT-AND-SEE (WS) & EXPECTED VALUE (EV)")

# 2.1 Giải Wait-and-See (WS)
Z_s_star = {}
for s in S:
    m_ws = pyo.ConcreteModel()
    m_ws.x = pyo.Var(J, within=pyo.NonNegativeReals)
    m_ws.y = pyo.Var(J, within=pyo.NonNegativeReals)
    m_ws.c1 = pyo.Constraint(expr=sum(m_ws.x[j] for j in J) <= budget_1)
    m_ws.c2 = pyo.Constraint(expr=sum(m_ws.y[j] for j in J) <= budget_2)
    m_ws.link = pyo.Constraint(expr=m_ws.y['AI'] <= 0.5 * m_ws.x['H'])

    m_ws.obj = pyo.Objective(expr=sum(beta[j]*m_ws.x[j] for j in J) + sum(beta_s[s, j]*m_ws.y[j] for j in J), sense=pyo.maximize)
    solver.solve(m_ws)
    Z_s_star[s] = pyo.value(m_ws.obj)

WS_value = sum(p[s] * Z_s_star[s] for s in S)
print(f"Giá trị kỳ vọng của Wait-and-See (WS) = {WS_value:,.2f}")
print("-" * 50)

# 2.2 Giải Expected Value (EV) Problem
beta_s_bar = {j: sum(p[s]*beta_s[s, j] for s in S) for j in J}

m_ev = pyo.ConcreteModel()
m_ev.x = pyo.Var(J, within=pyo.NonNegativeReals)
m_ev.y = pyo.Var(J, within=pyo.NonNegativeReals)
m_ev.c1 = pyo.Constraint(expr=sum(m_ev.x[j] for j in J) <= budget_1)
m_ev.c2 = pyo.Constraint(expr=sum(m_ev.y[j] for j in J) <= budget_2)
m_ev.link = pyo.Constraint(expr=m_ev.y['AI'] <= 0.5 * m_ev.x['H'])

m_ev.obj = pyo.Objective(expr=sum(beta[j]*m_ev.x[j] for j in J) + sum(beta_s_bar[j]*m_ev.y[j] for j in J), sense=pyo.maximize)
solver.solve(m_ev)
x_EV = {j: pyo.value(m_ev.x[j]) for j in J}

print("Quyết định First-stage kịch bản trung bình (x_EV):")
for j in J:
    print(f"  + Đầu tư {j}: {x_EV[j]:,.2f}")
print("\n=> NHẬN XÉT:")
print("- Quyết định x_EV thường có xu hướng cực đoan (dồn hết vốn vào 1 hoặc 2 hạng mục có lợi nhuận trung bình cao nhất) và bỏ qua rủi ro của các kịch bản xấu, khác biệt rõ so với chiến lược phân bổ đa dạng của x_SP.")


# =====================================================================
# CÂU 10.5.3: TÍNH VÀ PHÂN TÍCH VSS, EVPI
# =====================================================================
print("\n[3] CÂU 10.5.3 - TÍNH VSS VÀ EVPI")

# Tính EVPI
EVPI = WS_value - Z_SP

# Tính EEV (Fix x = x_EV vào mô hình SP gốc)
m_eev = create_base_model()
m_eev.obj = pyo.Objective(expr=sum(beta[j]*m_eev.x[j] for j in J) + sum(p[s]*sum(beta_s[s, j]*m_eev.y[s, j] for j in J) for s in S), sense=pyo.maximize)

m_eev.fix_x = pyo.ConstraintList()
for j in J:
    m_eev.fix_x.add(m_eev.x[j] == x_EV[j])

solver.solve(m_eev)
EEV = pyo.value(m_eev.obj)

# Tính VSS
VSS = Z_SP - EEV

print(f"Giá trị EEV = {EEV:,.2f}")
print(f"Giá trị VSS (Value of Stochastic Solution)     = {VSS:,.2f}")
print(f"Giá trị EVPI (Expected Value of Perf. Info) = {EVPI:,.2f}")
print("-" * 50)
print("=> Ý NGHĨA KINH TẾ:")
print("- VSS: Đại diện cho 'Giá trị của Giải pháp ngẫu nhiên'. Nó cho biết việc sử dụng mô hình SP (xét mọi kịch bản cùng lúc) sẽ giúp hệ thống thu lợi nhiều hơn bao nhiêu so với việc nhắm mắt làm theo kế hoạch trung bình (EV). VSS lớn cho thấy tính bất định rất quan trọng và không thể bỏ qua.")
print("- EVPI: Đại diện cho 'Giá trị của Thông tin hoàn hảo'. Đây là cái giá tối đa mà người ra quyết định sẵn sàng chi trả để biết trước kịch bản tương lai (ví dụ: mua báo cáo dự báo thị trường). Nó đo lường tổn thất tiềm tàng do sự thiếu vắng thông tin chắc chắn.")


# =====================================================================
# CÂU 10.5.4: ROBUST OPTIMIZATION (MINIMAX REGRET)
# =====================================================================
print("\n[4] CÂU 10.5.4 - ROBUST OPTIMIZATION (MINIMAX REGRET)")

m_rob = create_base_model()
m_rob.MaxRegret = pyo.Var(within=pyo.NonNegativeReals)

# Ràng buộc Minimax Regret
m_rob.regret_constr = pyo.ConstraintList()
for s in S:
    obj_s = sum(beta[j]*m_rob.x[j] for j in J) + sum(beta_s[s, j]*m_rob.y[s, j] for j in J)
    m_rob.regret_constr.add(Z_s_star[s] - obj_s <= m_rob.MaxRegret)

# Hàm mục tiêu: Minimize MaxRegret
m_rob.obj = pyo.Objective(expr=m_rob.MaxRegret, sense=pyo.minimize)
solver.solve(m_rob)

x_rob = {j: pyo.value(m_rob.x[j]) for j in J}
print(f"Max Regret tối ưu = {pyo.value(m_rob.MaxRegret):,.2f}")
print("Quyết định First-stage Robust (x_ROBUST):")
for j in J:
    print(f"  + Đầu tư {j}: {x_rob[j]:,.2f}")

print("\n=> NHẬN XÉT SỰ KHÁC BIỆT:")
print("- Mô hình SP tập trung tối đa hóa giá trị kỳ vọng nên ưu tiên phân bổ vào hạng mục mang lại lợi nhuận cao nhất tính theo xác suất (nghiêng về các kịch bản khả năng cao như s1, s2).")
print("- Mô hình Minimax Regret lại cực kỳ bảo thủ. Nó lo sợ nhất kịch bản xấu (s4) xảy ra gây 'tiếc nuối' lớn, nên x_ROBUST buộc phải chuyển hướng đầu tư mạnh hơn vào hạng mục H (Nhân lực) - nơi duy nhất có biên lợi nhuận tăng trong khủng hoảng (hệ số lên 1.10 ở kịch bản s4) để tự phòng vệ an toàn.")
```

```python
# =====================================================================
# CÀI ĐẶT MÔI TRƯỜNG & THƯ VIỆN (ẨN HOÀN TOÀN LOG CÀI ĐẶT)
# =====================================================================
!pip install -q stable-baselines3[extra] > /dev/null 2>&1

import gymnasium as gym
from gymnasium import spaces
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from stable_baselines3 import DQN
from stable_baselines3.common.vec_env import DummyVecEnv

# Thiết lập hiển thị đồ thị
plt.rcParams['figure.figsize'] = (10, 5)

# =====================================================================
# CÂU 11.3.1: CÀI ĐẶT ENVIRONMENT (VIETNAM ECONOMY ENV)
# =====================================================================
print("="*70)
print("1. KHỞI TẠO MÔI TRƯỜNG GYMNASIUM")
print("="*70)

class VietnamEconomyEnv(gym.Env):
    def __init__(self):
        super(VietnamEconomyEnv, self).__init__()
        # Action space: 5 chiến lược phân bổ
        self.action_space = spaces.Discrete(5)
        # Observation space: 4 chỉ số (Growth, D, AI, H), mỗi chỉ số 3 mức (0,1,2)
        self.observation_space = spaces.MultiDiscrete([3, 3, 3, 3])
        self.T = 10

        # Các chiến lược phân bổ (K, D, AI, H)
        self.allocation = {
            0: np.array([0.70, 0.10, 0.10, 0.10]), # Truyền thống
            1: np.array([0.40, 0.25, 0.15, 0.20]), # Cân bằng
            2: np.array([0.25, 0.45, 0.15, 0.15]), # Số hóa nhanh
            3: np.array([0.20, 0.20, 0.45, 0.15]), # AI dẫn dắt
            4: np.array([0.30, 0.20, 0.10, 0.40])  # Bao trùm
        }
        self.w = np.array([0.40, 0.25, 0.20, 0.15])

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.state = np.array([1, 1, 0, 1], dtype=np.int32)
        self.t = 0

        self.K = 27500.0
        self.D = 20.3
        self.AI = 86.0
        self.H = 30.0
        self.Y = (self.K**0.33) * (54.0**0.42) * (self.D**0.10) * (self.AI**0.08) * (self.H**0.07)
        return self.state, {}

    def step(self, action):
        a = self.allocation[action]
        budget = 1000.0
        Y_old = self.Y

        self.K += a[0] * budget
        self.D += a[1] * budget / 100.0
        self.AI += a[2] * budget / 20.0
        self.H += a[3] * budget / 200.0

        Y_new = (self.K**0.33) * (54.0**0.42) * (self.D**0.10) * (self.AI**0.08) * (self.H**0.07)
        self.Y = Y_new

        growth_rate = (Y_new - Y_old) / Y_old

        def discretize(val, thresh_low, thresh_high):
            if val < thresh_low: return 0
            elif val < thresh_high: return 1
            else: return 2

        g_idx = discretize(growth_rate, 0.03, 0.05)
        d_idx = discretize(self.D, 22.0, 26.0)
        ai_idx = discretize(self.AI, 95.0, 110.0)
        h_idx = discretize(self.H, 31.0, 33.0)

        self.state = np.array([g_idx, d_idx, ai_idx, h_idx], dtype=np.int32)
        self.t += 1

        reward = np.dot(self.w, self.state)
        done = self.t >= self.T

        return self.state, float(reward), done, False, {}

print("=> Đã khởi tạo môi trường 'VietnamEconomyEnv' thành công!\n")


# =====================================================================
# CÂU 11.3.2: HUẤN LUYỆN Q-LEARNING (TABULAR)
# =====================================================================
print("="*70)
print("2. HUẤN LUYỆN TABULAR Q-LEARNING")
print("="*70)

env = VietnamEconomyEnv()
Q = np.zeros((3, 3, 3, 3, 5))
alpha = 0.1
gamma = 0.95
n_episodes = 10000

returns_q = []

for ep in range(n_episodes):
    s, _ = env.reset()
    s = tuple(s)
    ep_return = 0
    eps = max(0.05, 1.0 - ep / 5000.0)

    while True:
        if np.random.rand() < eps:
            a = env.action_space.sample()
        else:
            a = np.argmax(Q[s])

        s_next, r, done, _, _ = env.step(a)
        s_next = tuple(s_next)
        ep_return += r

        Q[s][a] += alpha * (r + gamma * np.max(Q[s_next]) - Q[s][a])
        s = s_next

        if done:
            break
    returns_q.append(ep_return)

print(f"=> Đã huấn luyện thành công {n_episodes:,} episodes Q-Learning.\n")


# =====================================================================
# CÂU 11.3.3: TRÍCH XUẤT CHÍNH SÁCH TỐI ƯU
# =====================================================================
print("="*70)
print("3. TRÍCH XUẤT CHÍNH SÁCH TỐI ƯU TỪ BẢNG Q")
print("="*70)

def get_best_action(state):
    return np.argmax(Q[tuple(state)])

test_states = [
    [1, 1, 0, 1], # Trạng thái thực tế VN 2026
    [0, 0, 0, 0], # Mọi chỉ số đều thấp
    [2, 2, 2, 2], # Mọi chỉ số đều cao
    [0, 1, 2, 1], # Ngẫu nhiên 1
    [2, 0, 1, 0]  # Ngẫu nhiên 2
]

action_names = ["(0) Truyền thống", "(1) Cân bằng", "(2) Số hóa nhanh", "(3) AI dẫn dắt", "(4) Bao trùm"]

policy_results = []
for s in test_states:
    opt_a = get_best_action(s)
    policy_results.append({
        'Trạng thái (G, D, AI, H)': str(s),
        'Hành động tối ưu (Policy π*)': action_names[opt_a]
    })

df_policy = pd.DataFrame(policy_results)
print(df_policy.to_string(index=False))
print("\n")


# =====================================================================
# CÂU 11.3.4: SO SÁNH CÁC CHÍNH SÁCH VÀ VẼ ĐỒ THỊ
# =====================================================================
print("="*70)
print("4. ĐÁNH GIÁ VÀ SO SÁNH CHÍNH SÁCH (TEST 100 EPISODES)")
print("="*70)

def evaluate_policy(policy_type, n_eval=100):
    total_rewards = []
    for _ in range(n_eval):
        s, _ = env.reset()
        ep_ret = 0
        done = False
        while not done:
            if policy_type == 'q_learning':
                a = get_best_action(s)
            elif policy_type == 'rule_a1':
                a = 0
            elif policy_type == 'rule_a3':
                a = 2
            elif policy_type == 'random':
                a = env.action_space.sample()
            s, r, done, _, _ = env.step(a)
            ep_ret += r
        total_rewards.append(ep_ret)
    return np.mean(total_rewards)

# Lấy kết quả
q_avg = evaluate_policy('q_learning')
rule1_avg = evaluate_policy('rule_a1')
rule3_avg = evaluate_policy('rule_a3')
random_avg = evaluate_policy('random')

# Đưa vào DataFrame để in đẹp mắt
df_compare = pd.DataFrame({
    'Chính sách kiểm thử': ['1. Q-learning Optimal (π*)', '2. Rule a1 (Truyền thống)', '3. Rule a3 (Số hóa nhanh)', '4. Random (Ngẫu nhiên)'],
    'Phần thưởng TB (Return)': [q_avg, rule1_avg, rule3_avg, random_avg]
})
df_compare['Phần thưởng TB (Return)'] = df_compare['Phần thưởng TB (Return)'].map('{:.2f}'.format)
print(df_compare.to_string(index=False))
print("\n=> Chú ý đồ thị Learning Curve đang được vẽ bên dưới...\n")

# Vẽ Learning Curve
def moving_average(a, n=100):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

plt.figure(figsize=(10, 5))
plt.plot(moving_average(returns_q, n=100), label='Q-Learning (Moving Avg 100)', color='#1f77b4', linewidth=1.5)
plt.title('Đường Cong Học Tập (Learning Curve) của Q-Learning', fontsize=14, fontweight='bold')
plt.xlabel('Số chặng (Episodes)', fontsize=12)
plt.ylabel('Phần thưởng tích lũy (Return)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()


# =====================================================================
# CÂU 11.3.5: HUẤN LUYỆN MẠNG DQN BẰNG STABLE-BASELINES3
# =====================================================================
print("\n" + "="*70)
print("5. HUẤN LUYỆN DQN (DEEP Q-NETWORK) BẰNG STABLE-BASELINES3")
print("="*70)

vec_env = DummyVecEnv([lambda: VietnamEconomyEnv()])
model = DQN("MlpPolicy", vec_env, policy_kwargs=dict(net_arch=[64, 64]), learning_rate=1e-3, verbose=0)

print("=> Đang tiến hành huấn luyện Mạng Neural DQN (10,000 timesteps)...")
model.learn(total_timesteps=10000)

# Đánh giá 100 episodes
dqn_rewards = []
for _ in range(100):
    obs = vec_env.reset()
    ep_r = 0
    # DummyVecEnv yêu cầu lặp qua chính xác T steps
    for _ in range(10):
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, done, info = vec_env.step(action)
        ep_r += reward[0]
    dqn_rewards.append(ep_r)

dqn_avg = np.mean(dqn_rewards)
print(f"=> Hoàn tất! Phần thưởng trung bình của DQN (100 episodes test): {dqn_avg:.2f}")

print("-" * 50)
print("NHẬN XÉT SO SÁNH DQN vs TABULAR Q-LEARNING:")
if dqn_avg >= q_avg:
    print("- Mạng DQN đạt hiệu suất tương đương hoặc nhỉnh hơn Q-learning tabular.")
    print("- Bài toán này có không gian trạng thái khá nhỏ (3x3x3x3 = 81 trạng thái), nên Bảng Q truyền thống đã có thể ghi nhớ xuất sắc. Tuy nhiên, nếu mở rộng mô hình để nhận input là các giá trị thực tế liên tục (ví dụ GDP = 8044.4) thay vì các mức 0,1,2, thuật toán DQN sẽ thể hiện sức mạnh vượt trội nhờ khả năng xấp xỉ hàm (function approximation).")
else:
    print("- Với 10,000 timesteps, thuật toán DQN vẫn đang trong quá trình hội tụ trọng số mạng Neural và tạm thời chưa vượt qua được Q-learning Tabular.")
    print("- Nguyên nhân: Bài toán có không gian trạng thái nhỏ (3x3x3x3 = 81 trạng thái). Ở quy mô này, Bảng Q học thuộc lòng rất nhanh. Ngược lại, DQN cần nhiều dữ liệu hơn để huấn luyện Mạng Neural sâu. Bù lại, DQN sẽ vô đối nếu bài toán chuyển sang không gian trạng thái liên tục.")
```
